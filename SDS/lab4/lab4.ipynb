## Question 21: Verify Central Limit Theorem for non-normal population

```python
# Q21: Verify Central Limit Theorem for non-normal (uniform) population
print("="*60)
print("CENTRAL LIMIT THEOREM VERIFICATION - NON-NORMAL POPULATION")
print("="*60)

# Create a non-normal (uniform) population
non_normal_pop = np.random.uniform(0, 100, size=5000)
print(f"\nOriginal Population: Uniform Distribution (0-100)")
print(f"Population mean: {np.mean(non_normal_pop):.4f}")
print(f"Population std dev: {np.std(non_normal_pop):.4f}")

# Create sampling distributions with different sample sizes
sample_sizes = [10, 30, 50, 100]
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, n in enumerate(sample_sizes):
    sample_means_clt = []
    for i in range(1000):
        sample = np.random.choice(non_normal_pop, size=n, replace=True)
        sample_means_clt.append(np.mean(sample))
    
    sample_means_clt = np.array(sample_means_clt)
    
    # Plot histogram with normal distribution overlay
    axes[idx].hist(sample_means_clt, bins=30, edgecolor='black', alpha=0.7, 
                   color='skyblue', density=True, label='Sample Means Distribution')
    
    # Overlay normal distribution
    mu_samples = np.mean(sample_means_clt)
    sigma_samples = np.std(sample_means_clt)
    x = np.linspace(mu_samples - 4*sigma_samples, mu_samples + 4*sigma_samples, 100)
    axes[idx].plot(x, stats.norm.pdf(x, mu_samples, sigma_samples), 'r-', linewidth=2, 
                   label='Normal Distribution Fit')
    
    axes[idx].set_title(f'Sample Size = {n}, Mean = {mu_samples:.2f}, Std = {sigma_samples:.4f}')
    axes[idx].set_xlabel('Sample Mean')
    axes[idx].set_ylabel('Density')
    axes[idx].legend()
    axes[idx].grid(alpha=0.3)

plt.tight_layout()
plt.suptitle('Central Limit Theorem: Non-Normal Population', fontsize=14, y=1.00)
plt.show()

print("\n✓ As sample size increases, the distribution of sample means")
print("  approaches a normal distribution (Central Limit Theorem verified!)")
```

## Question 22: Verify Central Limit Theorem for normal population

```python
# Q22: Verify Central Limit Theorem for normal population
print("="*60)
print("CENTRAL LIMIT THEOREM VERIFICATION - NORMAL POPULATION")
print("="*60)

# Create a normal population
normal_pop = np.random.normal(loc=50, scale=15, size=5000)
print(f"\nOriginal Population: Normal Distribution (μ=50, σ=15)")
print(f"Population mean: {np.mean(normal_pop):.4f}")
print(f"Population std dev: {np.std(normal_pop):.4f}")

# Create sampling distributions with different sample sizes
sample_sizes = [10, 30, 50, 100]
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, n in enumerate(sample_sizes):
    sample_means_clt_norm = []
    for i in range(1000):
        sample = np.random.choice(normal_pop, size=n, replace=True)
        sample_means_clt_norm.append(np.mean(sample))
    
    sample_means_clt_norm = np.array(sample_means_clt_norm)
    
    # Plot histogram with normal distribution overlay
    axes[idx].hist(sample_means_clt_norm, bins=30, edgecolor='black', alpha=0.7, 
                   color='lightcoral', density=True, label='Sample Means Distribution')
    
    # Overlay normal distribution
    mu_samples = np.mean(sample_means_clt_norm)
    sigma_samples = np.std(sample_means_clt_norm)
    x = np.linspace(mu_samples - 4*sigma_samples, mu_samples + 4*sigma_samples, 100)
    axes[idx].plot(x, stats.norm.pdf(x, mu_samples, sigma_samples), 'b-', linewidth=2, 
                   label='Normal Distribution Fit')
    
    axes[idx].set_title(f'Sample Size = {n}, Mean = {mu_samples:.2f}, Std = {sigma_samples:.4f}')
    axes[idx].set_xlabel('Sample Mean')
    axes[idx].set_ylabel('Density')
    axes[idx].legend()
    axes[idx].grid(alpha=0.3)

plt.tight_layout()
plt.suptitle('Central Limit Theorem: Normal Population', fontsize=14, y=1.00)
plt.show()

print("\n✓ For normal population, the sample means distribution is normal")
print("  at all sample sizes (inherited from parent normal distribution)")
```

## Question 23: Verify CLT for Sampling Distribution of Sample Proportions

```python
# Q23: Verify CLT for Sampling Distribution of Sample Proportions
print("="*60)
print("CLT VERIFICATION FOR SAMPLING DISTRIBUTION OF PROPORTIONS")
print("="*60)

# Population parameters for smoking habits
population_size = 1000
true_proportion = 0.395  # True population proportion (39.5% smokers)

# Create binary population (1 = smoker, 0 = non-smoker)
population_binary = np.random.binomial(n=1, p=true_proportion, size=population_size)

print(f"\nPopulation Parameters:")
print(f"Population size: {population_size}")
print(f"True population proportion (p): {true_proportion}")
print(f"Observed population proportion: {np.mean(population_binary):.4f}")

# Sampling distribution of sample proportions with different sample sizes
sample_sizes_prop = [20, 50, 100, 200]
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

print(f"\nSampling Distribution Analysis:")
print("-" * 60)

for idx, n in enumerate(sample_sizes_prop):
    sample_proportions = []
    
    # Draw 1000 samples of size n and calculate proportion
    for i in range(1000):
        sample = np.random.choice(population_binary, size=n, replace=True)
        sample_prop = np.mean(sample)  # Proportion in this sample
        sample_proportions.append(sample_prop)
    
    sample_proportions = np.array(sample_proportions)
    
    # Theoretical values for sampling distribution of proportions
    p = true_proportion
    theoretical_mean = p
    theoretical_std = np.sqrt(p * (1 - p) / n)
    
    # Plot histogram with normal distribution overlay
    axes[idx].hist(sample_proportions, bins=30, edgecolor='black', alpha=0.7, 
                   color='lightyellow', density=True, label='Sample Proportions')
    
    # Overlay normal distribution
    x = np.linspace(theoretical_mean - 4*theoretical_std, 
                   theoretical_mean + 4*theoretical_std, 100)
    axes[idx].plot(x, stats.norm.pdf(x, theoretical_mean, theoretical_std), 
                  'r-', linewidth=2, label=f'Normal(μ={theoretical_mean:.3f}, σ={theoretical_std:.4f})')
    
    axes[idx].set_title(f'Sample Size = {n}')
    axes[idx].set_xlabel('Sample Proportion')
    axes[idx].set_ylabel('Density')
    axes[idx].legend(fontsize=9)
    axes[idx].grid(alpha=0.3)
    
    print(f"\nSample Size n = {n}:")
    print(f"  Mean of sample proportions: {np.mean(sample_proportions):.4f}")
    print(f"  Std dev of sample proportions: {np.std(sample_proportions):.4f}")
    print(f"  Theoretical std dev: {theoretical_std:.4f}")

plt.tight_layout()
plt.suptitle('Central Limit Theorem: Sampling Distribution of Proportions\n(Population p = 0.395)', 
             fontsize=14, y=1.00)
plt.show()

print("\n" + "="*60)
print("✓ Central Limit Theorem VERIFIED!")
print("As sample size increases, the sampling distribution of")
print("sample proportions approaches a normal distribution")
print(f"with mean = p = {true_proportion} and")
print(f"standard error = √[p(1-p)/n]")
print("="*60)
```

## Summary and Key Takeaways

### Random Number Generation
- **Integer Random Numbers**: Use `random.randint()` or `np.random.randint()`
- **Floating Point Numbers**: Use `random.uniform()` or `np.random.uniform()`

### Population Analysis
- Create large populations of random data (integers, floats, or from distributions)
- Calculate descriptive statistics: mean, variance, standard deviation

### Sampling Distributions
- Draw random samples from populations with user-defined sizes
- Create distributions of sample statistics (sample means, proportions)
- Observe how sample statistics vary across repeated sampling

### Central Limit Theorem (CLT)
1. **For any population distribution**, the sampling distribution of the sample mean:
   - Approaches a normal distribution as sample size increases
   - Has mean equal to the population mean (μ)
   - Has standard deviation equal to σ/√n (standard error)

2. **For Proportions**, the sampling distribution of sample proportions:
   - Approaches normal distribution for large samples
   - Has mean equal to population proportion (p)
   - Has standard deviation equal to √[p(1-p)/n]

3. **Key Finding**: CLT holds true regardless of the original population distribution!

### Practical Applications
- Confidence intervals and hypothesis testing
- Quality control and process monitoring

- Risk assessment and financial modeling
