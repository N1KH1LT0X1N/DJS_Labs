## Theory

### What is a Confidence Interval?

A **Confidence Interval (CI)** is a range of values that is estimated to contain a population parameter (such as mean or proportion) with a certain level of confidence. 

**Key Concepts:**

1. **Confidence Level**: The probability that the true parameter lies within the estimated interval. Common levels are 90%, 95%, and 99%.

2. **Margin of Error**: The range above and below the sample statistic.

3. **Critical Value**: A value from the standard normal distribution (Z) or t-distribution (t) that corresponds to the confidence level.

4. **Standard Error**: The standard deviation of the sampling distribution.

### Types of Confidence Intervals

#### 1. **Confidence Interval for Population Mean (μ) - Known σ**

When population standard deviation is known, use the Z-distribution:

$$CI = \bar{x} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$$

Where:
- $\bar{x}$ = sample mean
- $Z_{\alpha/2}$ = critical Z value
- $\sigma$ = population standard deviation
- $n$ = sample size

#### 2. **Confidence Interval for Population Mean (μ) - Unknown σ**

When population standard deviation is unknown, use the t-distribution:

$$CI = \bar{x} \pm t_{\alpha/2} \cdot \frac{s}{\sqrt{n}}$$

Where:
- $s$ = sample standard deviation
- $t_{\alpha/2}$ = critical t value with $n-1$ degrees of freedom

#### 3. **Confidence Interval for Population Proportion (p)**

For proportions, use the normal distribution if np ≥ 5 and n(1-p) ≥ 5:

$$CI = \hat{p} \pm Z_{\alpha/2} \cdot \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

Where:
- $\hat{p}$ = sample proportion
- $n$ = sample size

## Problem 1: Standard Normal Distribution - Probability Calculations

**Problem Statement:**
If Z follows the standard normal distribution N(0,1), find:

**(i)** $P(Z < 1.96)$  
**(ii)** $P(Z > -1.65)$  
**(iii)** $P(-1.96 < Z < 1.96)$  
**(iv)** $P(|Z| < 2.58)$  
**(v)** $P(-0.5 < Z < 1.5)$  
**(vi)** $P(Z > 2)$

### Theory: Standard Normal Distribution

The standard normal distribution (also called Z-distribution):
- Has mean = 0 and standard deviation = 1
- Is the foundation for most statistical inference
- Used when population standard deviation is known or sample size is large (n ≥ 30)
- Can convert any normal distribution to standard normal using: $Z = \frac{X - \mu}{\sigma}$

```python
# Problem 1: Standard Normal Distribution Probability Calculations
print("="*70)
print("PROBLEM 1: Standard Normal Distribution - Probability Calculations")
print("="*70)

# Problem 1(i): P(Z < 1.96)
print("\n(i) P(Z < 1.96)")
z_val_i = 1.96
prob_i = stats.norm.cdf(z_val_i)
print(f"    Using CDF: P(Z < {z_val_i}) = {prob_i:.6f}")

# Problem 1(ii): P(Z > -1.65)
print("\n(ii) P(Z > -1.65)")
z_val_ii = -1.65
prob_ii = 1 - stats.norm.cdf(z_val_ii)
print(f"     P(Z > {z_val_ii}) = 1 - P(Z < {z_val_ii})")
print(f"     P(Z > {z_val_ii}) = 1 - {stats.norm.cdf(z_val_ii):.6f}")
print(f"     P(Z > {z_val_ii}) = {prob_ii:.6f}")

# Problem 1(iii): P(-1.96 < Z < 1.96)
print("\n(iii) P(-1.96 < Z < 1.96)")
z_lower_iii = -1.96
z_upper_iii = 1.96
prob_iii = stats.norm.cdf(z_upper_iii) - stats.norm.cdf(z_lower_iii)
print(f"      P({z_lower_iii} < Z < {z_upper_iii}) = P(Z < {z_upper_iii}) - P(Z < {z_lower_iii})")
print(f"      P({z_lower_iii} < Z < {z_upper_iii}) = {stats.norm.cdf(z_upper_iii):.6f} - {stats.norm.cdf(z_lower_iii):.6f}")
print(f"      P({z_lower_iii} < Z < {z_upper_iii}) = {prob_iii:.6f}")

# Problem 1(iv): P(|Z| < 2.58)
print("\n(iv) P(|Z| < 2.58)")
z_val_iv = 2.58
prob_iv = stats.norm.cdf(z_val_iv) - stats.norm.cdf(-z_val_iv)
print(f"     P(|Z| < {z_val_iv}) = P(-{z_val_iv} < Z < {z_val_iv})")
print(f"     P(|Z| < {z_val_iv}) = {stats.norm.cdf(z_val_iv):.6f} - {stats.norm.cdf(-z_val_iv):.6f}")
print(f"     P(|Z| < {z_val_iv}) = {prob_iv:.6f}")

# Problem 1(v): P(-0.5 < Z < 1.5)
print("\n(v) P(-0.5 < Z < 1.5)")
z_lower_v = -0.5
z_upper_v = 1.5
prob_v = stats.norm.cdf(z_upper_v) - stats.norm.cdf(z_lower_v)
print(f"    P({z_lower_v} < Z < {z_upper_v}) = P(Z < {z_upper_v}) - P(Z < {z_lower_v})")
print(f"    P({z_lower_v} < Z < {z_upper_v}) = {stats.norm.cdf(z_upper_v):.6f} - {stats.norm.cdf(z_lower_v):.6f}")
print(f"    P({z_lower_v} < Z < {z_upper_v}) = {prob_v:.6f}")

# Problem 1(vi): P(Z > 2)
print("\n(vi) P(Z > 2)")
z_val_vi = 2
prob_vi = 1 - stats.norm.cdf(z_val_vi)
print(f"     P(Z > {z_val_vi}) = 1 - P(Z < {z_val_vi})")
print(f"     P(Z > {z_val_vi}) = 1 - {stats.norm.cdf(z_val_vi):.6f}")
print(f"     P(Z > {z_val_vi}) = {prob_vi:.6f}")

print("\n" + "="*70)
print("SUMMARY - Problem 1:")
print("="*70)
print(f"(i)   P(Z < 1.96)       = {prob_i:.6f}")
print(f"(ii)  P(Z > -1.65)      = {prob_ii:.6f}")
print(f"(iii) P(-1.96 < Z < 1.96) = {prob_iii:.6f}")
print(f"(iv)  P(|Z| < 2.58)     = {prob_iv:.6f}")
print(f"(v)   P(-0.5 < Z < 1.5) = {prob_v:.6f}")
print(f"(vi)  P(Z > 2)          = {prob_vi:.6f}")
```

## Problem 2: Standard Normal Distribution - Inverse (Finding Z₀)

**Problem Statement:**
If Z follows standard normal distribution N(0,1), find value of $Z_0$ satisfying:

**(i)** $P(Z < Z_0) = 0.90$  
**(ii)** $P(Z < Z_0) = 0.95$  
**(iii)** $P(Z < Z_0) = 0.975$  
**(iv)** $P(|Z| < Z_0) = 0.95$  
**(v)** $P(|Z| < Z_0) = 0.99$  
**(vi)** $P(Z > Z_0) = 0.05$

### Theory: Inverse Standard Normal Distribution (Quantiles)

The inverse of the standard normal distribution is used to find Z-scores corresponding to specific probabilities.

**Common Applications:**
- Confidence intervals for population mean
- Hypothesis testing
- Finding critical values for tests
- Quality control and process monitoring

```python
# Problem 2: Inverse Standard Normal Distribution (Finding Z₀)
print("\n" + "="*70)
print("PROBLEM 2: Standard Normal Distribution - Finding Z₀ (Inverse)")
print("="*70)

# Problem 2(i): P(Z < Z₀) = 0.90
print("\n(i) P(Z < Z₀) = 0.90")
prob_i = 0.90
z0_i = stats.norm.ppf(prob_i)
print(f"    Z₀ = {z0_i:.6f}")
print(f"    Verification: P(Z < {z0_i:.6f}) = {stats.norm.cdf(z0_i):.6f}")

# Problem 2(ii): P(Z < Z₀) = 0.95
print("\n(ii) P(Z < Z₀) = 0.95")
prob_ii = 0.95
z0_ii = stats.norm.ppf(prob_ii)
print(f"     Z₀ = {z0_ii:.6f}")
print(f"     Verification: P(Z < {z0_ii:.6f}) = {stats.norm.cdf(z0_ii):.6f}")

# Problem 2(iii): P(Z < Z₀) = 0.975
print("\n(iii) P(Z < Z₀) = 0.975")
prob_iii = 0.975
z0_iii = stats.norm.ppf(prob_iii)
print(f"      Z₀ = {z0_iii:.6f}")
print(f"      Verification: P(Z < {z0_iii:.6f}) = {stats.norm.cdf(z0_iii):.6f}")

# Problem 2(iv): P(|Z| < Z₀) = 0.95
# For two-sided: P(|Z| < Z₀) = 0.95 means P(-Z₀ < Z < Z₀) = 0.95
# So P(Z < Z₀) = 0.975
print("\n(iv) P(|Z| < Z₀) = 0.95")
prob_iv_two_sided = 0.95
prob_iv_one_sided = 0.5 + prob_iv_two_sided / 2
z0_iv = stats.norm.ppf(prob_iv_one_sided)
print(f"     Two-sided: P(|Z| < Z₀) = 0.95")
print(f"     One-sided: P(Z < Z₀) = 0.5 + 0.95/2 = {prob_iv_one_sided}")
print(f"     Z₀ = {z0_iv:.6f}")
print(f"     Verification: P(|Z| < {z0_iv:.6f}) = {2*stats.norm.cdf(z0_iv) - 1:.6f}")

# Problem 2(v): P(|Z| < Z₀) = 0.99
print("\n(v) P(|Z| < Z₀) = 0.99")
prob_v_two_sided = 0.99
prob_v_one_sided = 0.5 + prob_v_two_sided / 2
z0_v = stats.norm.ppf(prob_v_one_sided)
print(f"    Two-sided: P(|Z| < Z₀) = 0.99")
print(f"    One-sided: P(Z < Z₀) = 0.5 + 0.99/2 = {prob_v_one_sided}")
print(f"    Z₀ = {z0_v:.6f}")
print(f"    Verification: P(|Z| < {z0_v:.6f}) = {2*stats.norm.cdf(z0_v) - 1:.6f}")

# Problem 2(vi): P(Z > Z₀) = 0.05, which means P(Z < Z₀) = 0.95
print("\n(vi) P(Z > Z₀) = 0.05")
prob_vi_upper = 0.05
prob_vi = 1 - prob_vi_upper  # P(Z < Z₀) = 0.95
z0_vi = stats.norm.ppf(prob_vi)
print(f"     If P(Z > Z₀) = 0.05, then P(Z < Z₀) = 0.95")
print(f"     Z₀ = {z0_vi:.6f}")
print(f"     Verification: P(Z > {z0_vi:.6f}) = {1 - stats.norm.cdf(z0_vi):.6f}")

print("\n" + "="*70)
print("SUMMARY - Problem 2:")
print("="*70)
print(f"(i)   P(Z < Z₀) = 0.90   → Z₀ = {z0_i:.6f}")
print(f"(ii)  P(Z < Z₀) = 0.95   → Z₀ = {z0_ii:.6f}")
print(f"(iii) P(Z < Z₀) = 0.975  → Z₀ = {z0_iii:.6f}")
print(f"(iv)  P(|Z| < Z₀) = 0.95 → Z₀ = {z0_iv:.6f}")
print(f"(v)   P(|Z| < Z₀) = 0.99 → Z₀ = {z0_v:.6f}")
print(f"(vi)  P(Z > Z₀) = 0.05   → Z₀ = {z0_vi:.6f}")
```

## Problem 3: Student's t-Distribution - Probability Calculations

If t follows student's t distribution, then find:

**(i)** $P(t < 1.5)$ with d.o.f. = 20  
**(ii)** $P(t > 0.5)$ with d.o.f. = 15  
**(iii)** $P(|t| < 1.5)$ with d.o.f. = 25  
**(iv)** $P(|t| > 0.5)$ with d.o.f. = 35  
**(v)** $P(-2.2 < t < 1)$ with d.o.f. = 42

### Theory: Student's t-Distribution

The t-distribution is a probability distribution used when:
- Dealing with small samples (n < 30)
- Population standard deviation is unknown
- Estimating population mean from sample data

**Properties:**
- Symmetric around 0 (like standard normal)
- Heavier tails than normal distribution
- Converges to normal distribution as degrees of freedom increase
- Defined by one parameter: degrees of freedom (df = n - 1)

```python
# Problem 3: t-Distribution Probability Calculations
print("="*70)
print("PROBLEM 3: Student's t-Distribution - Probability Calculations")
print("="*70)

# Problem 3(i): P(t < 1.5) with df = 20
print("\n(i) P(t < 1.5) with d.o.f. = 20")
df_i = 20
t_val_i = 1.5
prob_i = stats.t.cdf(t_val_i, df_i)
print(f"    Using CDF: P(t < {t_val_i}) = {prob_i:.6f}")

# Problem 3(ii): P(t > 0.5) with df = 15
print("\n(ii) P(t > 0.5) with d.o.f. = 15")
df_ii = 15
t_val_ii = 0.5
prob_ii = 1 - stats.t.cdf(t_val_ii, df_ii)
print(f"     P(t > {t_val_ii}) = 1 - P(t < {t_val_ii})")
print(f"     P(t > {t_val_ii}) = 1 - {stats.t.cdf(t_val_ii, df_ii):.6f}")
print(f"     P(t > {t_val_ii}) = {prob_ii:.6f}")

# Problem 3(iii): P(|t| < 1.5) with df = 25
print("\n(iii) P(|t| < 1.5) with d.o.f. = 25")
df_iii = 25
t_val_iii = 1.5
# P(|t| < 1.5) = P(-1.5 < t < 1.5) = P(t < 1.5) - P(t < -1.5)
prob_iii = stats.t.cdf(t_val_iii, df_iii) - stats.t.cdf(-t_val_iii, df_iii)
print(f"     P(|t| < {t_val_iii}) = P(-{t_val_iii} < t < {t_val_iii})")
print(f"     P(|t| < {t_val_iii}) = P(t < {t_val_iii}) - P(t < -{t_val_iii})")
print(f"     P(|t| < {t_val_iii}) = {stats.t.cdf(t_val_iii, df_iii):.6f} - {stats.t.cdf(-t_val_iii, df_iii):.6f}")
print(f"     P(|t| < {t_val_iii}) = {prob_iii:.6f}")

# Problem 3(iv): P(|t| > 0.5) with df = 35
print("\n(iv) P(|t| > 0.5) with d.o.f. = 35")
df_iv = 35
t_val_iv = 0.5
# P(|t| > 0.5) = 1 - P(|t| < 0.5)
prob_iv_abs_less = stats.t.cdf(t_val_iv, df_iv) - stats.t.cdf(-t_val_iv, df_iv)
prob_iv = 1 - prob_iv_abs_less
print(f"     P(|t| > {t_val_iv}) = 1 - P(|t| < {t_val_iv})")
print(f"     P(|t| > {t_val_iv}) = 1 - {prob_iv_abs_less:.6f}")
print(f"     P(|t| > {t_val_iv}) = {prob_iv:.6f}")

# Problem 3(v): P(-2.2 < t < 1) with df = 42
print("\n(v) P(-2.2 < t < 1) with d.o.f. = 42")
df_v = 42
t_lower_v = -2.2
t_upper_v = 1
prob_v = stats.t.cdf(t_upper_v, df_v) - stats.t.cdf(t_lower_v, df_v)
print(f"    P({t_lower_v} < t < {t_upper_v}) = P(t < {t_upper_v}) - P(t < {t_lower_v})")
print(f"    P({t_lower_v} < t < {t_upper_v}) = {stats.t.cdf(t_upper_v, df_v):.6f} - {stats.t.cdf(t_lower_v, df_v):.6f}")
print(f"    P({t_lower_v} < t < {t_upper_v}) = {prob_v:.6f}")

print("\n" + "="*70)
print("SUMMARY - Problem 3:")
print("="*70)
print(f"(i)   P(t < 1.5) with df=20    = {prob_i:.6f}")
print(f"(ii)  P(t > 0.5) with df=15    = {prob_ii:.6f}")
print(f"(iii) P(|t| < 1.5) with df=25  = {prob_iii:.6f}")
print(f"(iv)  P(|t| > 0.5) with df=35  = {prob_iv:.6f}")
print(f"(v)   P(-2.2 < t < 1) with df=42 = {prob_v:.6f}")
```

## Problem 4: Student's t-Distribution - Inverse (Finding t₀)

If t follows student's t distribution, find value of $t_0$ satisfying:

**(i)** $P(t < t_0) = 0.90$ with d.o.f. = 20  
**(ii)** $P(t < t_0) = 0.95$ with d.o.f. = 15  
**(iii)** $P(t < t_0) = 0.99$ with d.o.f. = 25  
**(iv)** $P(|t| < t_0) = 0.90$ with d.o.f. = 30  
**(v)** $P(|t| < t_0) = 0.95$ with d.o.f. = 42  
**(vi)** $P(|t| < t_0) = 0.99$ with d.o.f. = 10

### Theory: Inverse t-Distribution (Quantiles)

The inverse of the t-distribution is used to find the critical value $t_0$ such that a specified probability is achieved.

**For one-sided tests:** Find $t_0$ such that $P(t < t_0) = α$  
**For two-sided tests:** Find $t_0$ such that $P(|t| < t_0) = 1 - α$

```python
# Problem 4: Inverse t-Distribution (Finding t₀)
print("\n" + "="*70)
print("PROBLEM 4: Student's t-Distribution - Finding t₀ (Inverse)")
print("="*70)

# Problem 4(i): P(t < t₀) = 0.90 with df = 20
print("\n(i) P(t < t₀) = 0.90 with d.o.f. = 20")
df_i = 20
prob_i = 0.90
t0_i = stats.t.ppf(prob_i, df_i)
print(f"    t₀ = {t0_i:.6f}")
print(f"    Verification: P(t < {t0_i:.6f}) = {stats.t.cdf(t0_i, df_i):.6f}")

# Problem 4(ii): P(t < t₀) = 0.95 with df = 15
print("\n(ii) P(t < t₀) = 0.95 with d.o.f. = 15")
df_ii = 15
prob_ii = 0.95
t0_ii = stats.t.ppf(prob_ii, df_ii)
print(f"     t₀ = {t0_ii:.6f}")
print(f"     Verification: P(t < {t0_ii:.6f}) = {stats.t.cdf(t0_ii, df_ii):.6f}")

# Problem 4(iii): P(t < t₀) = 0.99 with df = 25
print("\n(iii) P(t < t₀) = 0.99 with d.o.f. = 25")
df_iii = 25
prob_iii = 0.99
t0_iii = stats.t.ppf(prob_iii, df_iii)
print(f"      t₀ = {t0_iii:.6f}")
print(f"      Verification: P(t < {t0_iii:.6f}) = {stats.t.cdf(t0_iii, df_iii):.6f}")

# Problem 4(iv): P(|t| < t₀) = 0.90 with df = 30
# For two-sided: P(|t| < t₀) = 0.90 means P(-t₀ < t < t₀) = 0.90
# So P(t < t₀) = 0.95
print("\n(iv) P(|t| < t₀) = 0.90 with d.o.f. = 30")
df_iv = 30
prob_iv_two_sided = 0.90
prob_iv_one_sided = 0.5 + prob_iv_two_sided / 2
t0_iv = stats.t.ppf(prob_iv_one_sided, df_iv)
print(f"     Two-sided: P(|t| < t₀) = 0.90")
print(f"     One-sided: P(t < t₀) = 0.5 + 0.90/2 = {prob_iv_one_sided}")
print(f"     t₀ = {t0_iv:.6f}")
print(f"     Verification: P(|t| < {t0_iv:.6f}) = {2*stats.t.cdf(t0_iv, df_iv) - 1:.6f}")

# Problem 4(v): P(|t| < t₀) = 0.95 with df = 42
print("\n(v) P(|t| < t₀) = 0.95 with d.o.f. = 42")
df_v = 42
prob_v_two_sided = 0.95
prob_v_one_sided = 0.5 + prob_v_two_sided / 2
t0_v = stats.t.ppf(prob_v_one_sided, df_v)
print(f"    Two-sided: P(|t| < t₀) = 0.95")
print(f"    One-sided: P(t < t₀) = 0.5 + 0.95/2 = {prob_v_one_sided}")
print(f"    t₀ = {t0_v:.6f}")
print(f"    Verification: P(|t| < {t0_v:.6f}) = {2*stats.t.cdf(t0_v, df_v) - 1:.6f}")

# Problem 4(vi): P(|t| < t₀) = 0.99 with df = 10
print("\n(vi) P(|t| < t₀) = 0.99 with d.o.f. = 10")
df_vi = 10
prob_vi_two_sided = 0.99
prob_vi_one_sided = 0.5 + prob_vi_two_sided / 2
t0_vi = stats.t.ppf(prob_vi_one_sided, df_vi)
print(f"     Two-sided: P(|t| < t₀) = 0.99")
print(f"     One-sided: P(t < t₀) = 0.5 + 0.99/2 = {prob_vi_one_sided}")
print(f"     t₀ = {t0_vi:.6f}")
print(f"     Verification: P(|t| < {t0_vi:.6f}) = {2*stats.t.cdf(t0_vi, df_vi) - 1:.6f}")

print("\n" + "="*70)
print("SUMMARY - Problem 4:")
print("="*70)
print(f"(i)   P(t < t₀) = 0.90 with df=20    → t₀ = {t0_i:.6f}")
print(f"(ii)  P(t < t₀) = 0.95 with df=15    → t₀ = {t0_ii:.6f}")
print(f"(iii) P(t < t₀) = 0.99 with df=25    → t₀ = {t0_iii:.6f}")
print(f"(iv)  P(|t| < t₀) = 0.90 with df=30  → t₀ = {t0_iv:.6f}")
print(f"(v)   P(|t| < t₀) = 0.95 with df=42  → t₀ = {t0_v:.6f}")
print(f"(vi)  P(|t| < t₀) = 0.99 with df=10  → t₀ = {t0_vi:.6f}")
```

## Problem 5: Snedecor's F-Distribution - Probability Calculations

If F follows Snedecor's F distribution, then find:

**(i)** $P(F < 1.5)$ with $df_1 = 5, df_2 = 14$  
**(ii)** $P(F > 2.5)$ with $df_1 = 15, df_2 = 14$  
**(iii)** $P(0.5 < F < 4.1)$ with $df_1 = 13, df_2 = 17$

### Theory: Snedecor's F-Distribution

The F-distribution is used for:
- Comparing variances of two populations
- ANOVA (Analysis of Variance) tests
- Testing the equality of multiple means

**Properties:**
- Always positive (F > 0)
- Right-skewed distribution
- Defined by two parameters: $df_1$ (numerator degrees of freedom) and $df_2$ (denominator degrees of freedom)
- Relation to other distributions: If $T \sim t(df)$, then $T^2 \sim F(1, df)$

```python
# Problem 5: F-Distribution Probability Calculations
print("\n" + "="*70)
print("PROBLEM 5: Snedecor's F-Distribution - Probability Calculations")
print("="*70)

# Problem 5(i): P(F < 1.5) with df1 = 5, df2 = 14
print("\n(i) P(F < 1.5) with df₁ = 5, df₂ = 14")
df1_i = 5
df2_i = 14
f_val_i = 1.5
prob_i = stats.f.cdf(f_val_i, df1_i, df2_i)
print(f"    Using CDF: P(F < {f_val_i}) = {prob_i:.6f}")

# Problem 5(ii): P(F > 2.5) with df1 = 15, df2 = 14
print("\n(ii) P(F > 2.5) with df₁ = 15, df₂ = 14")
df1_ii = 15
df2_ii = 14
f_val_ii = 2.5
prob_ii = 1 - stats.f.cdf(f_val_ii, df1_ii, df2_ii)
print(f"     P(F > {f_val_ii}) = 1 - P(F < {f_val_ii})")
print(f"     P(F > {f_val_ii}) = 1 - {stats.f.cdf(f_val_ii, df1_ii, df2_ii):.6f}")
print(f"     P(F > {f_val_ii}) = {prob_ii:.6f}")

# Problem 5(iii): P(0.5 < F < 4.1) with df1 = 13, df2 = 17
print("\n(iii) P(0.5 < F < 4.1) with df₁ = 13, df₂ = 17")
df1_iii = 13
df2_iii = 17
f_lower_iii = 0.5
f_upper_iii = 4.1
prob_iii = stats.f.cdf(f_upper_iii, df1_iii, df2_iii) - stats.f.cdf(f_lower_iii, df1_iii, df2_iii)
print(f"     P({f_lower_iii} < F < {f_upper_iii}) = P(F < {f_upper_iii}) - P(F < {f_lower_iii})")
print(f"     P({f_lower_iii} < F < {f_upper_iii}) = {stats.f.cdf(f_upper_iii, df1_iii, df2_iii):.6f} - {stats.f.cdf(f_lower_iii, df1_iii, df2_iii):.6f}")
print(f"     P({f_lower_iii} < F < {f_upper_iii}) = {prob_iii:.6f}")

print("\n" + "="*70)
print("SUMMARY - Problem 5:")
print("="*70)
print(f"(i)   P(F < 1.5) with df₁=5, df₂=14     = {prob_i:.6f}")
print(f"(ii)  P(F > 2.5) with df₁=15, df₂=14    = {prob_ii:.6f}")
print(f"(iii) P(0.5 < F < 4.1) with df₁=13, df₂=17 = {prob_iii:.6f}")
```

## Problem 7: Chi-Squared Distribution - Probability Calculations

If X follows $χ^2$-distribution, then find:

**(i)** $P(X < 1.5)$ with $df = 10$  
**(ii)** $P(X > 2.5)$ with $df = 5$  
**(iii)** $P(0.5 < X < 4.1)$ with $df = 2$

### Theory: Chi-Squared Distribution (χ²)

The chi-squared distribution is used for:
- Goodness-of-fit tests
- Independence tests (contingency tables)
- Confidence intervals for variance
- Hypothesis testing for variance in a population

**Properties:**
- Always positive (X > 0)
- Right-skewed distribution
- Defined by one parameter: degrees of freedom (df)
- Special case: $χ^2(df) = \Gamma(df/2, 2)$ where Γ is the Gamma distribution
- Mean = df, Variance = 2df

```python
# Problem 7: Chi-Squared Distribution Probability Calculations
print("\n" + "="*70)
print("PROBLEM 7: Chi-Squared (χ²) Distribution - Probability Calculations")
print("="*70)

# Problem 7(i): P(X < 1.5) with df = 10
print("\n(i) P(X < 1.5) with df = 10")
df_i = 10
x_val_i = 1.5
prob_i = stats.chi2.cdf(x_val_i, df_i)
print(f"    Using CDF: P(X < {x_val_i}) = {prob_i:.6f}")

# Problem 7(ii): P(X > 2.5) with df = 5
print("\n(ii) P(X > 2.5) with df = 5")
df_ii = 5
x_val_ii = 2.5
prob_ii = 1 - stats.chi2.cdf(x_val_ii, df_ii)
print(f"     P(X > {x_val_ii}) = 1 - P(X < {x_val_ii})")
print(f"     P(X > {x_val_ii}) = 1 - {stats.chi2.cdf(x_val_ii, df_ii):.6f}")
print(f"     P(X > {x_val_ii}) = {prob_ii:.6f}")

# Problem 7(iii): P(0.5 < X < 4.1) with df = 2
print("\n(iii) P(0.5 < X < 4.1) with df = 2")
df_iii = 2
x_lower_iii = 0.5
x_upper_iii = 4.1
prob_iii = stats.chi2.cdf(x_upper_iii, df_iii) - stats.chi2.cdf(x_lower_iii, df_iii)
print(f"      P({x_lower_iii} < X < {x_upper_iii}) = P(X < {x_upper_iii}) - P(X < {x_lower_iii})")
print(f"      P({x_lower_iii} < X < {x_upper_iii}) = {stats.chi2.cdf(x_upper_iii, df_iii):.6f} - {stats.chi2.cdf(x_lower_iii, df_iii):.6f}")
print(f"      P({x_lower_iii} < X < {x_upper_iii}) = {prob_iii:.6f}")

print("\n" + "="*70)
print("SUMMARY - Problem 7:")
print("="*70)
print(f"(i)   P(X < 1.5) with df=10       = {prob_i:.6f}")
print(f"(ii)  P(X > 2.5) with df=5        = {prob_ii:.6f}")
print(f"(iii) P(0.5 < X < 4.1) with df=2  = {prob_iii:.6f}")
```

## Problem 9: Confidence Interval for Population Mean (Known σ)

**Problem Statement:**
Construct a 95% confidence interval for population mean in an experiment that found the sample mean temperature for a certain city in August was 101.82°F, with a population standard deviation of 1.2. There were 6 samples in this experiment.

**Given:**
- Sample mean ($\bar{x}$) = 101.82
- Population standard deviation (σ) = 1.2
- Sample size (n) = 6
- Confidence level = 95%

```python
# Problem 9 Solution
print("="*60)
print("PROBLEM 9: CI for Population Mean (Known σ)")
print("="*60)

# Given data
sample_mean = 101.82
sigma = 1.2  # Population standard deviation
n = 6
confidence_level = 0.95
alpha = 1 - confidence_level

print(f"\nGiven Data:")
print(f"  Sample Mean (x̄) = {sample_mean}")
print(f"  Population Standard Deviation (σ) = {sigma}")
print(f"  Sample Size (n) = {n}")
print(f"  Confidence Level = {confidence_level*100}%")
print(f"  α = {alpha}")

# Find critical Z value for 95% confidence interval
# For 95% CI, we need Z at α/2 = 0.025
z_critical = stats.norm.ppf(1 - alpha/2)
print(f"\nCritical Z value (Z_α/2) = {z_critical:.4f}")

# Calculate Standard Error
standard_error = sigma / np.sqrt(n)
print(f"Standard Error (σ/√n) = {standard_error:.4f}")

# Calculate Margin of Error
margin_of_error = z_critical * standard_error
print(f"Margin of Error = {margin_of_error:.4f}")

# Calculate Confidence Interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"\n{'─'*60}")
print(f"CONFIDENCE INTERVAL:")
print(f"{'─'*60}")
print(f"CI = x̄ ± Z_α/2 × (σ/√n)")
print(f"CI = {sample_mean} ± {margin_of_error:.4f}")
print(f"CI = [{lower_bound:.4f}, {upper_bound:.4f}]")
print(f"{'─'*60}")
print(f"\nInterpretation:")
print(f"We are 95% confident that the true population mean temperature")
print(f"lies between {lower_bound:.4f}°F and {upper_bound:.4f}°F")
```

## Problem 10: Confidence Interval for Population Mean (Unknown σ)

**Problem Statement:**
Construct a 98% Confidence Interval for population mean based on the following data:
45, 55, 67, 45, 68, 79, 98, 87, 84, 82

**Note:** The population standard deviation is unknown, so we use the t-distribution with sample standard deviation.

```python
# Problem 10 Solution
print("\n" + "="*60)
print("PROBLEM 10: CI for Population Mean (Unknown σ)")
print("="*60)

# Given data
data = np.array([45, 55, 67, 45, 68, 79, 98, 87, 84, 82])
confidence_level = 0.98
alpha = 1 - confidence_level

print(f"\nGiven Data: {data}")

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Using unbiased sample standard deviation
n = len(data)

print(f"\nSample Statistics:")
print(f"  Sample Size (n) = {n}")
print(f"  Sample Mean (x̄) = {sample_mean:.4f}")
print(f"  Sample Std Dev (s) = {sample_std:.4f}")
print(f"  Degrees of Freedom (df) = {n-1}")
print(f"  Confidence Level = {confidence_level*100}%")
print(f"  α = {alpha}")

# Find critical t value for 98% confidence interval
# With df = 9 and α/2 = 0.01
df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)
print(f"\nCritical t value (t_α/2, df) = {t_critical:.4f}")

# Calculate Standard Error
standard_error = sample_std / np.sqrt(n)
print(f"Standard Error (s/√n) = {standard_error:.4f}")

# Calculate Margin of Error
margin_of_error = t_critical * standard_error
print(f"Margin of Error = {margin_of_error:.4f}")

# Calculate Confidence Interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"\n{'─'*60}")
print(f"CONFIDENCE INTERVAL:")
print(f"{'─'*60}")
print(f"CI = x̄ ± t_α/2 × (s/√n)")
print(f"CI = {sample_mean:.4f} ± {margin_of_error:.4f}")
print(f"CI = [{lower_bound:.4f}, {upper_bound:.4f}]")
print(f"{'─'*60}")
print(f"\nInterpretation:")
print(f"We are 98% confident that the true population mean")
print(f"lies between {lower_bound:.4f} and {upper_bound:.4f}")
```

## Problem 11: Confidence Interval for Population Proportion

**Problem Statement:**
510 people applied to the Bachelor's in Elementary Education program at Florida State College. Of those applicants, 57 were men. Find the 90% CI of the true proportion of men who applied to the program.

**Given:**
- Total applicants (n) = 510
- Number of men (x) = 57
- Confidence level = 90%

```python
# Problem 11 Solution
print("\n" + "="*60)
print("PROBLEM 11: CI for Population Proportion")
print("="*60)

# Given data
n = 510  # total sample size
x = 57   # number of men
confidence_level = 0.90
alpha = 1 - confidence_level

print(f"\nGiven Data:")
print(f"  Total Applicants (n) = {n}")
print(f"  Number of Men (x) = {x}")
print(f"  Confidence Level = {confidence_level*100}%")
print(f"  α = {alpha}")

# Calculate sample proportion
p_hat = x / n
print(f"\nSample Proportion (p̂) = {x}/{n} = {p_hat:.6f}")

# Check assumptions for normal approximation
np_check = n * p_hat
n_1_p_check = n * (1 - p_hat)
print(f"\nAssumptions Check:")
print(f"  n×p̂ = {np_check:.2f} (should be ≥ 5)")
print(f"  n×(1-p̂) = {n_1_p_check:.2f} (should be ≥ 5)")
print(f"  ✓ Assumptions satisfied - normal approximation is appropriate")

# Find critical Z value for 90% confidence interval
# For 90% CI, we need Z at α/2 = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)
print(f"\nCritical Z value (Z_α/2) = {z_critical:.4f}")

# Calculate Standard Error for proportion
standard_error = np.sqrt((p_hat * (1 - p_hat)) / n)
print(f"Standard Error = √[p̂(1-p̂)/n] = {standard_error:.6f}")

# Calculate Margin of Error
margin_of_error = z_critical * standard_error
print(f"Margin of Error = {margin_of_error:.6f}")

# Calculate Confidence Interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

print(f"\n{'─'*60}")
print(f"CONFIDENCE INTERVAL:")
print(f"{'─'*60}")
print(f"CI = p̂ ± Z_α/2 × √[p̂(1-p̂)/n]")
print(f"CI = {p_hat:.6f} ± {margin_of_error:.6f}")
print(f"CI = [{lower_bound:.6f}, {upper_bound:.6f}]")
print(f"\nIn Percentage Form:")
print(f"CI = [{lower_bound*100:.4f}%, {upper_bound*100:.4f}%]")
print(f"{'─'*60}")
print(f"\nInterpretation:")
print(f"We are 90% confident that the true proportion of men")
print(f"who applied to the program is between {lower_bound*100:.4f}% and {upper_bound*100:.4f}%")
```

## Key Concepts and Takeaways

### 1. **Z vs t Distribution**
- Use **Z-distribution** when population standard deviation (σ) is known or sample size is large (n > 30)
- Use **t-distribution** when population standard deviation is unknown and sample size is small (n < 30)
- t-distribution has heavier tails, resulting in wider confidence intervals (more conservative estimate)

### 2. **Effect of Confidence Level**
- **Higher confidence level** (e.g., 99% vs 95%) → Wider confidence interval
- **Lower confidence level** (e.g., 90% vs 95%) → Narrower confidence interval
- There's a trade-off between precision and confidence

### 3. **Effect of Sample Size**
- **Larger sample size** (n) → Smaller standard error → Narrower confidence interval
- **Smaller sample size** → Larger standard error → Wider confidence interval
- Sample size should be determined based on desired precision and confidence level

### 4. **Population vs Sample Parameters**
- Population parameter (μ, σ, p) are fixed but unknown
- Sample statistics (x̄, s, p̂) are used to estimate population parameters
- Confidence interval provides a range of plausible values for the true parameter

### 5. **Practical Interpretation**
- A 95% CI means: If we repeated the experiment 100 times, approximately 95 of those confidence intervals would contain the true parameter
- It does **NOT** mean the probability the parameter is in this specific interval is 95%

### 6. **Assumptions**
- Data should be approximately normally distributed (or n should be large enough)
- Samples should be independent
- For proportion CI: np ≥ 5 and n(1-p) ≥ 5 for normal approximation to be valid

## Key Relationships Between Distributions

1. **Normal Distribution** → Foundation for all statistical inference
2. **t-Distribution** → Emerges from: $t = \frac{\bar{X} - \mu}{S/\sqrt{n}}$
3. **F-Distribution** → Ratio of two chi-squared: $F = \frac{\chi^2_{df_1}/df_1}{\chi^2_{df_2}/df_2}$
4. **Chi-Squared Distribution** → Square of standard normal: $\chi^2(1) = Z^2$

## When to Use Each Distribution

| Distribution | When to Use | Key Parameters |
|---|---|---|
| **Z (Normal)** | Large samples (n ≥ 30), σ known | μ = 0, σ = 1 |
| **t** | Small samples (n < 30), σ unknown | df = n - 1 |
| **F** | Comparing variances, ANOVA | df₁, df₂ |
| **χ²** | Goodness-of-fit, independence tests | df |
