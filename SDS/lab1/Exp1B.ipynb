## Theory

Measures of central tendency and dispersion are fundamental statistical tools used to summarize and understand data.

### Measures of Central Tendency:
- **Mean**: The arithmetic average of all values
- **Median**: The middle value when data are arranged in order
- **Mode**: The most frequently occurring value

### Measures of Dispersion:
- **Range**: The difference between maximum and minimum values
- **Variance**: The average of squared deviations from the mean
- **Standard Deviation**: The square root of variance
- **Coefficient of Variation**: Ratio of standard deviation to mean
- **Interquartile Range**: The difference between Q3 and Q1

While central tendency shows where the data are centered, dispersion reveals the degree of variability or consistency within the dataset.

### Q1. Find arithmetic mean of 20, 2, 7, 1, 34

```python
# Q1: Find arithmetic mean
data_q1 = [20, 2, 7, 1, 34]
print(f"Data: {data_q1}")

# Method 1: Using manual calculation
mean_manual = sum(data_q1) / len(data_q1)
print(f"\nArithmetic Mean (manual): {mean_manual}")

# Method 2: Using statistics library
mean_stats = statistics.mean(data_q1)
print(f"Arithmetic Mean (statistics): {mean_stats}")

# Method 3: Using numpy
mean_numpy = np.mean(data_q1)
print(f"Arithmetic Mean (numpy): {mean_numpy}")
```

### Q4. Find weighted average

```python
# Q4: Weighted Average
# Example: Values with corresponding weights
values = [60, 75, 85, 90]
weights = [0.2, 0.2, 0.3, 0.3]  # Total should be 1

print(f"Values: {values}")
print(f"Weights: {weights}")

# Calculate weighted average
weighted_avg = np.average(values, weights=weights)
print(f"\nWeighted Average: {weighted_avg}")

# Alternative calculation
weighted_avg_manual = sum(v * w for v, w in zip(values, weights)) / sum(weights)
print(f"Weighted Average (manual): {weighted_avg_manual}")
```

### Q6. Calculate geometric mean for each column, row, and entire data

```python
# Q6: Geometric Mean
from scipy import stats

data_q6 = np.array([
    [1, 3, 27],
    [3, 4, 6],
    [7, 6, 3],
    [3, 6, 8]
])

print("Data Matrix:")
print(data_q6)

# Geometric mean of entire data
geom_mean_all = stats.gmean(data_q6.flatten())
print(f"\nGeometric Mean (entire data): {geom_mean_all}")

# Geometric mean for each column
geom_mean_cols = np.array([stats.gmean(data_q6[:, i]) for i in range(data_q6.shape[1])])
print(f"\nGeometric Mean (each column): {geom_mean_cols}")

# Geometric mean for each row
geom_mean_rows = np.array([stats.gmean(data_q6[i, :]) for i in range(data_q6.shape[0])])
print(f"\nGeometric Mean (each row): {geom_mean_rows}")
```

### Q7. Calculate harmonic mean for 1, 3, 5, 7, 9

```python
# Q7: Harmonic Mean
from scipy import stats

data_q7 = [1, 3, 5, 7, 9]
print(f"Data: {data_q7}")

# Harmonic mean using scipy
harmonic_mean = stats.hmean(data_q7)
print(f"\nHarmonic Mean: {harmonic_mean}")

# Alternative calculation (H = n / (1/x1 + 1/x2 + ... + 1/xn))
n = len(data_q7)
harmonic_mean_manual = n / sum(1/x for x in data_q7)
print(f"Harmonic Mean (manual): {harmonic_mean_manual}")
```

### Q9. Solar heating systems - Mean, Mode, Median analysis

```python
# Q9: Solar heating systems heat-storage capacity
# Sample data for 28 systems (in days)
solar_data = [1.5, 2.0, 2.3, 2.5, 2.7, 2.8, 3.0, 3.2, 3.5, 3.8, 
              4.0, 4.2, 4.5, 4.8, 5.0, 5.2, 5.5, 5.8, 6.0, 6.2,
              6.5, 6.8, 7.0, 7.2, 7.5, 7.8, 8.0, 8.5]

print(f"Heat-storage capacity data (n={len(solar_data)}):")
print(solar_data)

# (a) Compute the mean
mean_solar = np.mean(solar_data)
print(f"\n(a) Mean: {mean_solar}")

# (b) Compute the mode
counter_solar = Counter(solar_data)
mode_count = max(counter_solar.values())
modes_solar = [k for k, v in counter_solar.items() if v == mode_count]
print(f"\n(b) Mode: {modes_solar[0] if modes_solar else 'No mode (all values appear once)'}")

# (c) Compute the median
median_solar = np.median(solar_data)
print(f"\n(c) Median: {median_solar}")

# (d) Which measure best reflects the data?
print(f"\n(d) Analysis:")
print(f"    Mean = {mean_solar:.2f}")
print(f"    Median = {median_solar:.2f}")
print(f"    The median is slightly less affected by extreme values and provides")
print(f"    a better representation of the central tendency of this dataset.")
```

### Q10. Write a function for percentile calculation

```python
# Q10: Percentile Function
def calculate_percentile(data, percentile):
    """
    Calculate the percentile value of a dataset
    
    Args:
        data: List or array of values
        percentile: Percentile value (0-100)
    
    Returns:
        The value at the given percentile
    """
    sorted_data = sorted(data)
    index = (percentile / 100) * (len(sorted_data) - 1)
    
    # Linear interpolation
    lower_index = int(index)
    upper_index = min(lower_index + 1, len(sorted_data) - 1)
    weight = index - lower_index
    
    return sorted_data[lower_index] * (1 - weight) + sorted_data[upper_index] * weight

# Test data
data_q10 = [30, 40, 72, 83, 25, 10, 50, 90, 60, 15, 5, 9, 34, 23, 67, 80, 67, 45]
print(f"Data: {data_q10}")
print(f"Sorted: {sorted(data_q10)}")

# Calculate percentiles
percentiles = [30, 50, 75, 90]
print(f"\nPercentile Values:")
for p in percentiles:
    value = calculate_percentile(data_q10, p)
    print(f"  {p}th percentile: {value:.2f}")

# Verification using numpy
print(f"\nVerification using NumPy:")
for p in percentiles:
    value = np.percentile(data_q10, p)
    print(f"  {p}th percentile: {value:.2f}")
```

### Q12. Calculate standard deviation of plant heights

```python
# Q12: Standard Deviation
heights = [51, 38, 79, 46, 57]  # in cm
print(f"Plant heights (cm): {heights}")

# Mean
mean_height = np.mean(heights)
print(f"\nMean: {mean_height}")

# Variance (population)
variance_pop = np.var(heights)
print(f"\nVariance (population): {variance_pop}")

# Variance (sample)
variance_sample = np.var(heights, ddof=1)
print(f"Variance (sample): {variance_sample}")

# Standard Deviation (population)
std_pop = np.std(heights)
print(f"\nStandard Deviation (population): {std_pop}")

# Standard Deviation (sample)
std_sample = np.std(heights, ddof=1)
print(f"Standard Deviation (sample): {std_sample}")

# Manual calculation
deviations = [(h - mean_height)**2 for h in heights]
std_manual = np.sqrt(sum(deviations) / (len(heights) - 1))
print(f"\nStandard Deviation (sample, manual calculation): {std_manual}")
```

### Q14. Chef's decision on tomato weights

```python
# Q14: Chef's Tomato Quality Decision
tomato_weights = [6.3, 7.2, 7.3, 8.1, 7.8, 6.8, 7.5, 7.8, 7.2, 7.5, 8.1, 8.2,
                  8.0, 7.4, 7.6, 7.7, 7.6, 7.4, 7.5, 8.4, 7.4, 7.6, 6.2, 7.4]

print(f"Tomato weights (ounces): {tomato_weights}")
print(f"Number of tomatoes: {len(tomato_weights)}")

# Calculate statistics
mean_weight = np.mean(tomato_weights)
std_weight = np.std(tomato_weights, ddof=1)

print(f"\nAverage weight: {mean_weight:.2f} ounces")
print(f"Standard deviation: {std_weight:.4f} ounces")

# Chef's criteria
print(f"\nChef's Criteria:")
print(f"  Required average: 7.5 ounces")
print(f"  Maximum acceptable std deviation: 0.5 ounce")

# Decision
print(f"\nDecision:")
if mean_weight == 7.5 and std_weight < 0.5:
    print(f"  ✓ ACCEPT - Average is {mean_weight:.2f} (correct) and std dev is {std_weight:.4f} (< 0.5)")
else:
    if mean_weight != 7.5:
        print(f"  ✗ REJECT - Average {mean_weight:.2f} ≠ 7.5")
    if std_weight >= 0.5:
        print(f"  ✗ REJECT - Std deviation {std_weight:.4f} ≥ 0.5")
```

### Q15. Compare variability between two training programs

```python
# Q15: Comparison of Training Program Variability
print("Training Program Comparison:")
print("="*50)

# Group 1: Program A
mean_a = 32.11
variance_a = 68.09
std_a = np.sqrt(variance_a)
cv_a = (std_a / mean_a) * 100  # Coefficient of Variation

print(f"\nProgram A (Group 1):")
print(f"  Mean training time: {mean_a} hours")
print(f"  Variance: {variance_a}")
print(f"  Standard Deviation: {std_a:.4f}")
print(f"  Coefficient of Variation: {cv_a:.2f}%")

# Group 2: Program B
mean_b = 19.75
variance_b = 71.14
std_b = np.sqrt(variance_b)
cv_b = (std_b / mean_b) * 100

print(f"\nProgram B (Group 2):")
print(f"  Mean training time: {mean_b} hours")
print(f"  Variance: {variance_b}")
print(f"  Standard Deviation: {std_b:.4f}")
print(f"  Coefficient of Variation: {cv_b:.2f}%")

# Comparison
print(f"\n{'='*50}")
print(f"Conclusion:")
if cv_a < cv_b:
    print(f"Program A has LESS relative variability ({cv_a:.2f}% < {cv_b:.2f}%)")
    print(f"Program A is more consistent and recommended.")
else:
    print(f"Program B has LESS relative variability ({cv_b:.2f}% < {cv_a:.2f}%)")
    print(f"Program B is more consistent and recommended.")
```

### Q17. Comparative box plot analysis of day and evening classes

```python
# Q17: Box Plot Analysis
day_class = [99, 56, 78, 55.5, 32, 90, 80, 81, 56, 59, 45, 77, 84.5, 84, 70, 72, 68, 32, 79, 90]
evening_class = [98, 78, 68, 83, 81, 89, 88, 76, 65, 45, 98, 90, 80, 84.5, 85, 79, 78, 98, 90, 79, 81, 25.5]

print("Day Class Scores:")
print(sorted(day_class))
print(f"Number of students: {len(day_class)}")

print("\nEvening Class Scores:")
print(sorted(evening_class))
print(f"Number of students: {len(evening_class)}")

# Calculate statistics for Day Class
day_min = np.min(day_class)
day_max = np.max(day_class)
day_median = np.median(day_class)
day_q1 = np.percentile(day_class, 25)
day_q3 = np.percentile(day_class, 75)

print(f"\nDay Class Statistics:")
print(f"  Minimum: {day_min}")
print(f"  Maximum: {day_max}")
print(f"  Q1 (25th percentile): {day_q1}")
print(f"  Median (50th percentile): {day_median}")
print(f"  Q3 (75th percentile): {day_q3}")

# Calculate statistics for Evening Class
evening_min = np.min(evening_class)
evening_max = np.max(evening_class)
evening_median = np.median(evening_class)
evening_q1 = np.percentile(evening_class, 25)
evening_q3 = np.percentile(evening_class, 75)

print(f"\nEvening Class Statistics:")
print(f"  Minimum: {evening_min}")
print(f"  Maximum: {evening_max}")
print(f"  Q1 (25th percentile): {evening_q1}")
print(f"  Median (50th percentile): {evening_median}")
print(f"  Q3 (75th percentile): {evening_q3}")

# Create box plots
fig, ax = plt.subplots(figsize=(10, 6))
data_to_plot = [day_class, evening_class]
ax.boxplot(data_to_plot, labels=['Day Class', 'Evening Class'])
ax.set_ylabel('Test Scores')
ax.set_title('Comparative Box Plot: Day vs Evening Class')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Calculate percentage distributions
print(f"\n--- Percentage Analysis for Day Class ---")
print(f"Min to Q1: {(day_q1 - day_min) / (day_max - day_min) * 100:.1f}%")
print(f"Q1 to Median: {(day_median - day_q1) / (day_max - day_min) * 100:.1f}%")
print(f"Median to Q3: {(day_q3 - day_median) / (day_max - day_min) * 100:.1f}%")
print(f"Q3 to Max: {(day_max - day_q3) / (day_max - day_min) * 100:.1f}%")
print(f"Q1 to Q3 (middle 50%): {(day_q3 - day_q1) / (day_max - day_min) * 100:.1f}%")

print(f"\n--- Percentage Analysis for Evening Class ---")
print(f"Min to Q1: {(evening_q1 - evening_min) / (evening_max - evening_min) * 100:.1f}%")
print(f"Q1 to Median: {(evening_median - evening_q1) / (evening_max - evening_min) * 100:.1f}%")
print(f"Median to Q3: {(evening_q3 - evening_median) / (evening_max - evening_min) * 100:.1f}%")
print(f"Q3 to Max: {(evening_max - evening_q3) / (evening_max - evening_min) * 100:.1f}%")
print(f"Q1 to Q3 (middle 50%): {(evening_q3 - evening_q1) / (evening_max - evening_min) * 100:.1f}%")

# Compare spreads
day_iqr = day_q3 - day_q1
evening_iqr = evening_q3 - evening_q1

print(f"\n--- Comparison ---")
print(f"Day Class IQR: {day_iqr}")
print(f"Evening Class IQR: {evening_iqr}")
if day_iqr > evening_iqr:
    print(f"Day class has WIDER spread in middle 50% (more variability)")
else:
    print(f"Evening class has WIDER spread in middle 50% (more variability)")
```

## Conclusion

We have successfully studied and implemented measures of central tendency and dispersion in Python using various libraries (NumPy, SciPy, and Statistics). Key takeaways:

### Measures of Central Tendency:
- **Mean**: Best for normally distributed data
- **Median**: Best for skewed distributions or data with outliers
- **Mode**: Best for categorical data or identifying the most frequent value

### Measures of Dispersion:
- **Range**: Simple but affected by outliers
- **Variance**: Measures average squared deviations
- **Standard Deviation**: Square root of variance, in same units as data
- **Coefficient of Variation**: Relative variability (useful for comparing different datasets)
- **Interquartile Range**: Robust measure unaffected by outliers

### Practical Applications:
- Understanding data distribution for better decision making
- Comparing datasets with different scales
- Identifying outliers and anomalies
- Quality control and statistical analysis

These statistical measures are fundamental tools for data analysis and are widely used in data science, business analytics, and research.
