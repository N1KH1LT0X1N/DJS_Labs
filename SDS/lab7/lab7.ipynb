# Theory Questions

## Question 1: Which statistical test is commonly used when the sample size is small (n < 30)?

**Answer:** The **t-test (Student's t-test)** is the most commonly used statistical test for small sample sizes (n < 30). When the population standard deviation is unknown and the sample size is small, the t-distribution (which has heavier tails than the normal distribution) is more appropriate than the normal distribution for hypothesis testing and constructing confidence intervals.

## Question 2: What is the main purpose of performing a t-test for small samples?

**Answer:** The main purposes of performing a t-test for small samples are:
- To test hypotheses about population means when the sample size is small and population standard deviation is unknown
- To compare means between two groups (independent or paired samples)
- To account for the uncertainty introduced by having only a small sample, using the t-distribution instead of the normal distribution
- To make valid statistical inferences with limited data
- To determine if observed differences between sample means are statistically significant or due to chance

## Question 1: Test if average age is 30

Given 11 ages, check whether average age is 30 or not.

**Data:** [28, 32, 29, 31, 27, 35, 30, 33, 26, 30, 28]

**Solution:**
- **Null Hypothesis (H₀):** μ = 30 (average age is 30)
- **Alternative Hypothesis (H₁):** μ ≠ 30 (average age is not 30)
- **Test:** One-sample t-test (two-tailed)
- **Significance Level:** α = 0.05

```python
# Question 1
ages = np.array([28, 32, 29, 31, 27, 35, 30, 33, 26, 30, 28])
mu_0 = 30
alpha = 0.05

# Calculate statistics
n = len(ages)
sample_mean = np.mean(ages)
sample_std = np.std(ages, ddof=1)  # Sample standard deviation
sample_sem = sample_std / np.sqrt(n)  # Standard error of the mean

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(ages, mu_0)

# Degrees of freedom
df = n - 1

# Critical t-value
t_critical = stats.t.ppf(1 - alpha/2, df)

print("\n" + "="*60)
print("QUESTION 1: Test if Average Age is 30")
print("="*60)
print(f"Sample size (n): {n}")
print(f"Sample mean: {sample_mean:.4f}")
print(f"Sample standard deviation: {sample_std:.4f}")
print(f"Standard error of mean: {sample_sem:.4f}")
print(f"\nHypothesis:")
print(f"  H₀: μ = 30")
print(f"  H₁: μ ≠ 30 (Two-tailed test)")
print(f"  α = {alpha}")
print(f"\nDegrees of freedom: {df}")
print(f"Critical t-value (±): {t_critical:.4f}")
print(f"\nTest Results:")
print(f"  t-statistic: {t_statistic:.4f}")
print(f"  p-value: {p_value:.4f}")
print(f"\nConclusion:")
if abs(t_statistic) > t_critical:
    print(f"  Since |t| = {abs(t_statistic):.4f} > {t_critical:.4f}, we REJECT H₀")
    print(f"  p-value = {p_value:.4f} < {alpha}")
    print(f"  The average age is SIGNIFICANTLY different from 30.")
else:
    print(f"  Since |t| = {abs(t_statistic):.4f} < {t_critical:.4f}, we FAIL TO REJECT H₀")
    print(f"  p-value = {p_value:.4f} ≥ {alpha}")
    print(f"  The average age is NOT significantly different from 30.")
```

## Question 2: Test if average math score is less than 50

Test the hypothesis that average score in math is less than 50 from the scores given below.

**Data:** [10, 20, 35, 50, 28, 40, 55, 18, 16, 55, 30, 25, 65, 18, 43, 23, 21, 20, 19, 70]

**Solution:**
- **Null Hypothesis (H₀):** μ = 50 (average score is 50)
- **Alternative Hypothesis (H₁):** μ < 50 (average score is less than 50)
- **Test:** One-sample t-test (left-tailed)
- **Significance Level:** α = 0.05

```python
# Question 2
math_scores = np.array([10, 20, 35, 50, 28, 40, 55, 18, 16, 55, 30, 25, 65, 18, 43, 23, 21, 20, 19, 70])
mu_0 = 50
alpha = 0.05

# Calculate statistics
n = len(math_scores)
sample_mean = np.mean(math_scores)
sample_std = np.std(math_scores, ddof=1)  # Sample standard deviation
sample_sem = sample_std / np.sqrt(n)  # Standard error of the mean

# Perform one-sample t-test (two-tailed, then adjust for one-tailed)
t_statistic_two_tailed, p_value_two_tailed = stats.ttest_1samp(math_scores, mu_0)
t_statistic = t_statistic_two_tailed

# For left-tailed test: if t is negative, p_value = p_value_two_tailed/2
# if t is positive, p_value = 1 - p_value_two_tailed/2
if t_statistic < 0:
    p_value = p_value_two_tailed / 2
else:
    p_value = 1 - p_value_two_tailed / 2

# Degrees of freedom
df = n - 1

# Critical t-value for left-tailed test
t_critical = stats.t.ppf(alpha, df)

print("\n" + "="*60)
print("QUESTION 2: Test if Average Math Score is Less than 50")
print("="*60)
print(f"Sample size (n): {n}")
print(f"Sample mean: {sample_mean:.4f}")
print(f"Sample standard deviation: {sample_std:.4f}")
print(f"Standard error of mean: {sample_sem:.4f}")
print(f"\nHypothesis:")
print(f"  H₀: μ = 50")
print(f"  H₁: μ < 50 (Left-tailed test)")
print(f"  α = {alpha}")
print(f"\nDegrees of freedom: {df}")
print(f"Critical t-value: {t_critical:.4f}")
print(f"\nTest Results:")
print(f"  t-statistic: {t_statistic:.4f}")
print(f"  p-value (one-tailed): {p_value:.4f}")
print(f"\nConclusion:")
if t_statistic < t_critical:
    print(f"  Since t = {t_statistic:.4f} < {t_critical:.4f}, we REJECT H₀")
    print(f"  p-value = {p_value:.4f} < {alpha}")
    print(f"  The average math score is SIGNIFICANTLY less than 50.")
else:
    print(f"  Since t = {t_statistic:.4f} ≥ {t_critical:.4f}, we FAIL TO REJECT H₀")
    print(f"  p-value = {p_value:.4f} ≥ {alpha}")
    print(f"  The average math score is NOT significantly less than 50.")
```

## Question 3: Two-sample t-test with Combined Populations

Consider two populations following Poisson distribution with mean age 35 years and 10 years respectively. The size of population 1 is 150000 and that of population 2 is 100000. Combine both of them to make one population. Two samples are taken from Gujarat state with size 30 and 20 and mean value 30 and 10 respectively. Combine both samples to make one sample. Test the hypothesis that mean age of Gujarat is equal to mean age of the country.

**Solution:**
- **Population Parameters:** 
  - Pop 1: μ₁ = 35, n₁ = 150000
  - Pop 2: μ₂ = 10, n₂ = 100000
  - Combined population mean: μ = (150000×35 + 100000×10)/(150000+100000) = 24
- **Sample Parameters:**
  - Sample 1: n₁ = 30, x̄₁ = 30
  - Sample 2: n₂ = 20, x̄₂ = 10
  - Combined sample mean: x̄ = (30×30 + 20×10)/(30+20) = 22
- **Null Hypothesis (H₀):** μ_sample = μ_population = 24
- **Alternative Hypothesis (H₁):** μ_sample ≠ 24
- **Test:** One-sample t-test on combined sample

```python
# Question 3
# Population parameters
pop1_mean = 35
pop1_size = 150000
pop2_mean = 10
pop2_size = 100000

# Combined population mean
combined_pop_mean = (pop1_size * pop1_mean + pop2_size * pop2_mean) / (pop1_size + pop2_size)

# Sample parameters
sample1_size = 30
sample1_mean = 30
sample2_size = 20
sample2_mean = 10

# Combined sample mean
combined_sample_mean = (sample1_size * sample1_mean + sample2_size * sample2_mean) / (sample1_size + sample2_size)

# Create combined sample data (using means repeated)
combined_sample = np.concatenate([
    np.full(sample1_size, sample1_mean),
    np.full(sample2_size, sample2_mean)
])

# For calculation purposes, assume sample variance based on the data
# Since we only have means, we'll estimate using reasonable variance
sample_variance = 25  # Estimated variance
sample_std = np.std(combined_sample, ddof=1)

n = len(combined_sample)
mu_0 = combined_pop_mean
alpha = 0.05

# Perform t-test
t_statistic, p_value = stats.ttest_1samp(combined_sample, mu_0)

df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)

print("\n" + "="*60)
print("QUESTION 3: Combined Population and Sample Means")
print("="*60)
print(f"\nPopulation Data:")
print(f"  Population 1: mean = {pop1_mean}, size = {pop1_size}")
print(f"  Population 2: mean = {pop2_mean}, size = {pop2_size}")
print(f"  Combined population mean: μ = {combined_pop_mean:.4f}")
print(f"\nSample Data:")
print(f"  Sample 1: n₁ = {sample1_size}, x̄₁ = {sample1_mean}")
print(f"  Sample 2: n₂ = {sample2_size}, x̄₂ = {sample2_mean}")
print(f"  Combined sample mean: x̄ = {combined_sample_mean:.4f}")
print(f"\nHypothesis:")
print(f"  H₀: Mean age of Gujarat = Mean age of country ({mu_0:.4f})")
print(f"  H₁: Means are not equal (Two-tailed test)")
print(f"  α = {alpha}")
print(f"\nTest Statistics:")
print(f"  Combined sample size (n): {n}")
print(f"  Sample mean: {np.mean(combined_sample):.4f}")
print(f"  Sample std dev: {sample_std:.4f}")
print(f"  Degrees of freedom: {df}")
print(f"  Critical t-value (±): {t_critical:.4f}")
print(f"\nTest Results:")
print(f"  t-statistic: {t_statistic:.4f}")
print(f"  p-value: {p_value:.4f}")
print(f"\nConclusion:")
if abs(t_statistic) > t_critical:
    print(f"  Since |t| = {abs(t_statistic):.4f} > {t_critical:.4f}, we REJECT H₀")
    print(f"  The mean age of Gujarat is significantly different from the country mean.")
else:
    print(f"  Since |t| = {abs(t_statistic):.4f} ≤ {t_critical:.4f}, we FAIL TO REJECT H₀")
    print(f"  The mean age of Gujarat is not significantly different from the country mean.")
```

## Question 4: Two-sample t-test - Independent Samples

Check the hypothesis that mean values of temperature for week 1 and week 2 are the same.

**Data:**
- Week 1: [30, 31, 29, 28, 32, 33, 30]
- Week 2: [32, 34, 33, 31, 30, 29, 28]

**Solution:**
- **Null Hypothesis (H₀):** μ₁ = μ₂ (mean temperatures are equal)
- **Alternative Hypothesis (H₁):** μ₁ ≠ μ₂ (mean temperatures are different)
- **Test:** Two-sample t-test (independent samples)
- **Significance Level:** α = 0.05
- **Hint:** Perform the two-sample t-test

```python
# Question 4
week1 = np.array([30, 31, 29, 28, 32, 33, 30])
week2 = np.array([32, 34, 33, 31, 30, 29, 28])
alpha = 0.05

# Calculate statistics for both samples
n1 = len(week1)
n2 = len(week2)
mean1 = np.mean(week1)
mean2 = np.mean(week2)
std1 = np.std(week1, ddof=1)
std2 = np.std(week2, ddof=1)

# Perform Levene's test to check for equal variances
levene_stat, levene_p = stats.levene(week1, week2)

# Perform two-sample t-test (assuming unequal variances)
t_statistic, p_value = stats.ttest_ind(week1, week2, equal_var=False)

# Degrees of freedom (Welch's approximation)
df = ((std1**2/n1 + std2**2/n2)**2) / ((std1**2/n1)**2/(n1-1) + (std2**2/n2)**2/(n2-1))

# Critical t-value
t_critical = stats.t.ppf(1 - alpha/2, df)

print("\n" + "="*60)
print("QUESTION 4: Two-sample t-test - Temperature Comparison")
print("="*60)
print(f"\nWeek 1 Statistics:")
print(f"  Sample size (n₁): {n1}")
print(f"  Mean (x̄₁): {mean1:.4f}")
print(f"  Standard deviation (s₁): {std1:.4f}")
print(f"\nWeek 2 Statistics:")
print(f"  Sample size (n₂): {n2}")
print(f"  Mean (x̄₂): {mean2:.4f}")
print(f"  Standard deviation (s₂): {std2:.4f}")
print(f"\nLevene's Test for Equal Variances:")
print(f"  Test statistic: {levene_stat:.4f}")
print(f"  p-value: {levene_p:.4f}")
if levene_p > alpha:
    print(f"  Conclusion: Variances are equal (p > {alpha})")
else:
    print(f"  Conclusion: Variances are not equal (p ≤ {alpha})")
print(f"\nHypothesis:")
print(f"  H₀: μ₁ = μ₂ (Mean temperatures are equal)")
print(f"  H₁: μ₁ ≠ μ₂ (Mean temperatures are different)")
print(f"  α = {alpha}")
print(f"\nTwo-Sample t-test Results (Welch's t-test):")
print(f"  Degrees of freedom: {df:.4f}")
print(f"  Critical t-value (±): {t_critical:.4f}")
print(f"  t-statistic: {t_statistic:.4f}")
print(f"  p-value: {p_value:.4f}")
print(f"\nConclusion:")
if abs(t_statistic) > t_critical:
    print(f"  Since |t| = {abs(t_statistic):.4f} > {t_critical:.4f}, we REJECT H₀")
    print(f"  p-value = {p_value:.4f} < {alpha}")
    print(f"  The mean temperatures are SIGNIFICANTLY different between week 1 and week 2.")
else:
    print(f"  Since |t| = {abs(t_statistic):.4f} ≤ {t_critical:.4f}, we FAIL TO REJECT H₀")
    print(f"  p-value = {p_value:.4f} ≥ {alpha}")
    print(f"  The mean temperatures are NOT significantly different between week 1 and week 2.")
```

## Question 5: Paired t-test - Blood Pressure Before and After Treatment

Consider patient's data with blood pressure values before and after treatment. Check the effectiveness of treatment.

**Solution:**
- **Null Hypothesis (H₀):** μ_difference = 0 (Treatment has no effect)
- **Alternative Hypothesis (H₁):** μ_difference ≠ 0 (Treatment has an effect)
- **Test:** Paired t-test
- **Significance Level:** α = 0.05
- **Hint:** Paired t-test

```python
# Question 5 - Paired t-test
# Load blood pressure data from CSV file
bp_data = pd.read_csv('blood_pressure.csv')

# Display the data
print("\n" + "="*60)
print("QUESTION 5: Paired t-test - Blood Pressure Before/After Treatment")
print("="*60)
print("\nBlood Pressure Data (First 10 rows):")
print(bp_data.head(10))

# Get before and after measurements
before = bp_data.iloc[:, 3].values  # bp_before column
after = bp_data.iloc[:, 4].values   # bp_after column

# Calculate differences
differences = after - before

n = len(differences)
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)
se_diff = std_diff / np.sqrt(n)
alpha = 0.05

# Perform paired t-test
t_statistic, p_value = stats.ttest_rel(after, before)

df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)

print(f"\nStatistics:")
print(f"  Number of patients (n): {n}")
print(f"  Mean before treatment: {np.mean(before):.4f}")
print(f"  Mean after treatment: {np.mean(after):.4f}")
print(f"  Mean difference (after - before): {mean_diff:.4f}")
print(f"  Std dev of differences: {std_diff:.4f}")
print(f"  Standard error of differences: {se_diff:.4f}")
print(f"\nHypothesis:")
print(f"  H₀: μ_difference = 0 (Treatment has no effect)")
print(f"  H₁: μ_difference ≠ 0 (Treatment has an effect)")
print(f"  α = {alpha}")
print(f"\nPaired t-test Results:")
print(f"  Degrees of freedom: {df}")
print(f"  Critical t-value (±): {t_critical:.4f}")
print(f"  t-statistic: {t_statistic:.4f}")
print(f"  p-value: {p_value:.4f}")
print(f"\nConclusion:")
if abs(t_statistic) > t_critical:
    print(f"  Since |t| = {abs(t_statistic):.4f} > {t_critical:.4f}, we REJECT H₀")
    print(f"  p-value = {p_value:.4f} < {alpha}")
    print(f"  The treatment IS EFFECTIVE in changing blood pressure.")
else:
    print(f"  Since |t| = {abs(t_statistic):.4f} ≤ {t_critical:.4f}, we FAIL TO REJECT H₀")
    print(f"  p-value = {p_value:.4f} ≥ {alpha}")
    print(f"  The treatment is NOT significantly effective in changing blood pressure.")
```

## Question 6: Confidence Interval for Population Variance

Construct a 95% Confidence Interval for population variance based on the following data:

**Data:** [40.2, 32.8, 38.2, 43.5, 47.6, 36.6, 38.4, 45.5, 44.4, 40.3, 34.6, 55.6, 50.9, 38.9, 37.8, 46.8, 43.6, 39.5, 49.9, 34.2]

**Solution:**
- **Confidence Level:** 95% (α = 0.05)
- **Test:** Chi-square distribution
- **Formula:** CI = [(n-1)s²/χ²(α/2, n-1), (n-1)s²/χ²(1-α/2, n-1)]
  where s² is sample variance and χ² values are from chi-square distribution

```python
# Question 6 - Confidence Interval for Population Variance
data = np.array([40.2, 32.8, 38.2, 43.5, 47.6, 36.6, 38.4, 45.5, 44.4, 40.3, 
                 34.6, 55.6, 50.9, 38.9, 37.8, 46.8, 43.6, 39.5, 49.9, 34.2])

n = len(data)
mean = np.mean(data)
sample_variance = np.var(data, ddof=1)  # Sample variance
sample_std = np.std(data, ddof=1)  # Sample standard deviation
confidence_level = 0.95
alpha = 1 - confidence_level
df = n - 1

# Chi-square critical values
chi2_upper = stats.chi2.ppf(1 - alpha/2, df)
chi2_lower = stats.chi2.ppf(alpha/2, df)

# Confidence interval for variance
ci_lower_variance = (df * sample_variance) / chi2_upper
ci_upper_variance = (df * sample_variance) / chi2_lower

# Confidence interval for standard deviation
ci_lower_std = np.sqrt(ci_lower_variance)
ci_upper_std = np.sqrt(ci_upper_variance)

print("\n" + "="*60)
print("QUESTION 6: 95% Confidence Interval for Population Variance")
print("="*60)
print(f"\nData Summary:")
print(f"  Sample size (n): {n}")
print(f"  Sample mean: {mean:.4f}")
print(f"  Sample variance (s²): {sample_variance:.4f}")
print(f"  Sample standard deviation (s): {sample_std:.4f}")
print(f"\nConfidence Level: {confidence_level*100}% (α = {alpha})")
print(f"\nDegrees of freedom: {df}")
print(f"\nChi-square Critical Values:")
print(f"  χ²(α/2, {df}) = χ²({alpha/2}, {df}) = {chi2_lower:.4f}")
print(f"  χ²(1-α/2, {df}) = χ²({1-alpha/2}, {df}) = {chi2_upper:.4f}")
print(f"\nFormula: CI = [(n-1)s²/χ²(α/2), (n-1)s²/χ²(1-α/2)]")
print(f"\n95% Confidence Interval for Population Variance:")
print(f"  Lower limit: ({df} × {sample_variance:.4f}) / {chi2_upper:.4f} = {ci_lower_variance:.4f}")
print(f"  Upper limit: ({df} × {sample_variance:.4f}) / {chi2_lower:.4f} = {ci_upper_variance:.4f}")
print(f"\n  CI for Variance (σ²): [{ci_lower_variance:.4f}, {ci_upper_variance:.4f}]")
print(f"\n95% Confidence Interval for Population Standard Deviation:")
print(f"  CI for Std Dev (σ): [{ci_lower_std:.4f}, {ci_upper_std:.4f}]")
print(f"\nInterpretation:")
print(f"  We are 95% confident that the population variance lies between {ci_lower_variance:.4f} and {ci_upper_variance:.4f}")
print(f"  We are 95% confident that the population standard deviation lies between {ci_lower_std:.4f} and {ci_upper_std:.4f}")
```

## Question 7: F-test for Equality of Variances

Two samples having 21 and 9 observations have variances given as 16 and 8 respectively. Test the hypothesis that the first population variance is greater than the second one at 0.05 and 0.01 level of significance.

**Solution:**
- **Sample 1:** n₁ = 21, s₁² = 16
- **Sample 2:** n₂ = 9, s₂² = 8
- **Null Hypothesis (H₀):** σ₁² = σ₂² (variances are equal)
- **Alternative Hypothesis (H₁):** σ₁² > σ₂² (first variance is greater)
- **Test:** F-test (right-tailed)
- **Significance Levels:** α = 0.05 and α = 0.01
- **Formula:** F = s₁²/s₂² with df₁ = n₁-1, df₂ = n₂-1

```python
# Question 7 - F-test for Equality of Variances
n1 = 21
var1 = 16
std1 = np.sqrt(var1)

n2 = 9
var2 = 8
std2 = np.sqrt(var2)

df1 = n1 - 1
df2 = n2 - 1

# Calculate F-statistic
f_statistic = var1 / var2

# Test at two significance levels
alpha_values = [0.05, 0.01]

print("\n" + "="*60)
print("QUESTION 7: F-test for Equality of Variances")
print("="*60)
print(f"\nSample Data:")
print(f"  Sample 1: n₁ = {n1}, s₁² = {var1}")
print(f"  Sample 2: n₂ = {n2}, s₂² = {var2}")
print(f"\nDegrees of Freedom:")
print(f"  df₁ = n₁ - 1 = {df1}")
print(f"  df₂ = n₂ - 1 = {df2}")
print(f"\nHypothesis:")
print(f"  H₀: σ₁² = σ₂² (variances are equal)")
print(f"  H₁: σ₁² > σ₂² (first variance is greater - right-tailed test)")
print(f"\nF-test Statistic:")
print(f"  F = s₁²/s₂² = {var1}/{var2} = {f_statistic:.4f}")

results_summary = []
for alpha in alpha_values:
    # Critical F-value for right-tailed test
    f_critical = stats.f.ppf(1 - alpha, df1, df2)
    
    # p-value for right-tailed test
    p_value = 1 - stats.f.cdf(f_statistic, df1, df2)
    
    results_summary.append((alpha, f_critical, p_value))
    
    print(f"\n{'─'*60}")
    print(f"At α = {alpha}:")
    print(f"  Critical F-value: F({df1}, {df2}, {1-alpha}) = {f_critical:.4f}")
    print(f"  p-value (right-tailed): {p_value:.4f}")
    print(f"\n  Decision:")
    if f_statistic > f_critical:
        print(f"    Since F = {f_statistic:.4f} > {f_critical:.4f}, we REJECT H₀")
        print(f"    The first population variance is SIGNIFICANTLY GREATER than the second.")
    else:
        print(f"    Since F = {f_statistic:.4f} ≤ {f_critical:.4f}, we FAIL TO REJECT H₀")
        print(f"    We cannot conclude that the first variance is greater than the second.")

print(f"\n\\nFinal Conclusion:")
alpha_005_critical, alpha_001_critical = results_summary[0][1], results_summary[1][1]
if f_statistic > alpha_005_critical:
    print(f"  At α = 0.05: The first population variance IS significantly greater.")
else:
    print(f"  At α = 0.05: The first population variance is NOT significantly greater.")
if f_statistic > alpha_001_critical:
    print(f"  At α = 0.01: The first population variance IS significantly greater.")
else:
    print(f"  At α = 0.01: The first population variance is NOT significantly greater.")
```

## Summary of Tests Performed:

1. **One-sample t-test (Two-tailed):** Used to test if a single sample mean differs significantly from a hypothesized population mean (Question 1)

2. **One-sample t-test (One-tailed):** Used to test if a sample mean is significantly less than or greater than a hypothesized value (Question 2)

## Key Concepts:

- **t-distribution:** Used for small samples (n < 30) when population standard deviation is unknown
- **Degrees of Freedom:** Critical in determining appropriate critical values for small sample tests
- **p-values:** Provide evidence against null hypothesis; smaller values indicate stronger evidence
- **Significance Level (α):** Determines the probability of Type I error (rejecting true null hypothesis)
- **Chi-square Distribution:** Used for inference about variances
- **F-distribution:** Used for comparing variances of two populations

These statistical tests are essential for analyzing small samples where the normal distribution approximation may not be appropriate.
