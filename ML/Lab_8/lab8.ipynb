# Lab 8: Association Rule Mining

## Aim
Explore Association Rule Mining with variations on different datasets.

## Theory
Association Rule Mining is a machine learning method meant to discover interesting relationships, frequent patterns, associations, or correlations among a set of items in transactional or relational data.

### Key Concepts:
- **Support**: The proportion of transactions that contain the itemset.
- **Confidence**: Indicates the number of times the if-then statements are found true.
- **Lift**: Ratio of observed support to expected support if X and Y were independent.

### Algorithms:
- **Apriori**: Uses breadth-first search and generates candidate itemsets
- **FP-Growth**: Uses prefix-tree structure for efficient mining

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
```

## Data Preparation

Preparing the transaction dataset for association rule mining:

```python
# Original transaction dataset
transactions = [
    ['Milk','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'], 
    ['Dill','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'], 
    ['Milk','Apple','Kidney Beans','Eggs'], 
    ['Milk','Unicorn','Corn','Kidney Beans','Yogurt'], 
    ['Corn','Onion','Onion','Kidney Beans','Ice cream','Eggs']
]

print("Transaction Dataset:")
for i, transaction in enumerate(transactions, 1):
    print(f"Transaction {i}: {transaction}")
print(f"\nTotal transactions: {len(transactions)}")
```

```python
# Convert transactions to one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

print("\nOne-Hot Encoded Transaction Data:")
print(df_encoded)
print(f"\nShape: {df_encoded.shape}")
print(f"Items: {list(df_encoded.columns)}")
```

## Task 1: Frequent Itemsets with Apriori (min_support = 0.6)

Generate frequent itemsets using the Apriori algorithm with a minimum support threshold of 0.6.

```python
# Task 1: Apriori with min_support = 0.6
min_support_1 = 0.6
frequent_itemsets_1 = apriori(df_encoded, min_support=min_support_1, use_colnames=True)

# Sort by support in descending order
frequent_itemsets_1_sorted = frequent_itemsets_1.sort_values('support', ascending=False)

print(f"Task 1: Frequent Itemsets (min_support = {min_support_1})")
print("=" * 80)
print(f"\nTotal frequent itemsets found: {len(frequent_itemsets_1_sorted)}\n")

# Display results in a formatted way
for idx, row in frequent_itemsets_1_sorted.iterrows():
    itemset = list(row['itemsets'])
    support = row['support']
    print(f"Itemset: {itemset}")
    print(f"Support: {support:.4f} ({support*100:.2f}%)")
    print("-" * 40)

# Store for later comparison
frequent_itemsets_06 = frequent_itemsets_1_sorted.copy()
```

## Task 2: Frequent Itemsets with Higher Threshold (min_support = 0.7) and Comparison

Modify the minimum support threshold to 0.7, regenerate frequent itemsets, and compare with Task 1.

```python
# Task 2: Apriori with min_support = 0.7
min_support_2 = 0.7
frequent_itemsets_2 = apriori(df_encoded, min_support=min_support_2, use_colnames=True)

# Sort by support in descending order
frequent_itemsets_2_sorted = frequent_itemsets_2.sort_values('support', ascending=False)

print(f"Task 2: Frequent Itemsets (min_support = {min_support_2})")
print("=" * 80)
print(f"\nTotal frequent itemsets found: {len(frequent_itemsets_2_sorted)}\n")

# Display results
for idx, row in frequent_itemsets_2_sorted.iterrows():
    itemset = list(row['itemsets'])
    support = row['support']
    print(f"Itemset: {itemset}")
    print(f"Support: {support:.4f} ({support*100:.2f}%)")
    print("-" * 40)

# Store for later comparison
frequent_itemsets_07 = frequent_itemsets_2_sorted.copy()
```

```python
# Comparison between Task 1 and Task 2
print("\n" + "=" * 80)
print("COMPARISON: Task 1 (min_support=0.6) vs Task 2 (min_support=0.7)")
print("=" * 80)

print(f"\nTask 1 (min_support=0.6):")
print(f"  - Total itemsets: {len(frequent_itemsets_06)}")
print(f"  - Itemsets found:")
for idx, row in frequent_itemsets_06.iterrows():
    itemset = list(row['itemsets'])
    support = row['support']
    print(f"    {itemset} → {support:.4f}")

print(f"\nTask 2 (min_support=0.7):")
print(f"  - Total itemsets: {len(frequent_itemsets_07)}")
print(f"  - Itemsets found:")
for idx, row in frequent_itemsets_07.iterrows():
    itemset = list(row['itemsets'])
    support = row['support']
    print(f"    {itemset} → {support:.4f}")

print(f"\nKey Observations:")
print(f"  1. Number of itemsets decreased from {len(frequent_itemsets_06)} to {len(frequent_itemsets_07)}")
print(f"     (decrease of {len(frequent_itemsets_06) - len(frequent_itemsets_07)} itemsets)")
print(f"  2. Increasing the support threshold filters out less frequent itemsets")
print(f"  3. Higher support means only items that appear together more frequently are retained")
print(f"  4. This makes the mining more conservative and focused on stronger patterns")
```

## Task 3: Association Rules Generation

Generate association rules from the frequent itemsets with confidence and lift metrics.

```python
# Task 3: Generate association rules from frequent itemsets (using min_support=0.6)
min_confidence = 0.0  # Start with all rules
rules = association_rules(frequent_itemsets_06, metric="confidence", min_threshold=min_confidence)

# Add additional metrics for better analysis
if len(rules) > 0:
    rules = rules.sort_values('lift', ascending=False)
    
    print("Task 3: Association Rules (from itemsets with min_support=0.6)")
    print("=" * 100)
    print(f"\nTotal association rules generated: {len(rules)}\n")
    
    # Display with proper formatting
    for idx, row in rules.iterrows():
        antecedent = list(row['antecedents'])
        consequent = list(row['consequents'])
        confidence = row['confidence']
        lift = row['lift']
        support = row['support']
        
        print(f"Rule {idx + 1}:")
        print(f"  {antecedent} → {consequent}")
        print(f"  Support: {support:.4f} ({support*100:.2f}%)")
        print(f"  Confidence: {confidence:.4f} ({confidence*100:.2f}%)")
        print(f"  Lift: {lift:.4f}")
        print("-" * 100)
    
    # Store for later use
    rules_06 = rules.copy()
else:
    print("No rules found with the given threshold.")
    rules_06 = pd.DataFrame()
```

## Task 4: Impact of Different Confidence Thresholds

Explore how different confidence thresholds affect the generated association rules.

```python
# Task 4: Explore different confidence thresholds
confidence_thresholds = [0.0, 0.3, 0.6, 0.8, 1.0]

print("Task 4: Impact of Different Confidence Thresholds")
print("=" * 100)
print(f"\nAnalyzing rules generated with different confidence thresholds:\n")

results_summary = []

for threshold in confidence_thresholds:
    rules_temp = association_rules(frequent_itemsets_06, metric="confidence", min_threshold=threshold)
    num_rules = len(rules_temp)
    
    results_summary.append({
        'Confidence Threshold': f"{threshold:.1f}",
        'Number of Rules': num_rules
    })
    
    print(f"\nConfidence Threshold: {threshold:.1f}")
    print(f"  Number of rules generated: {num_rules}")
    
    if num_rules > 0:
        avg_confidence = rules_temp['confidence'].mean()
        avg_lift = rules_temp['lift'].mean()
        max_lift = rules_temp['lift'].max()
        min_lift = rules_temp['lift'].min()
        
        print(f"  Average confidence: {avg_confidence:.4f}")
        print(f"  Average lift: {avg_lift:.4f}")
        print(f"  Lift range: [{min_lift:.4f}, {max_lift:.4f}]")
        
        print(f"\n  Rules:")
        rules_temp_sorted = rules_temp.sort_values('lift', ascending=False)
        for idx, row in rules_temp_sorted.head(5).iterrows():
            antecedent = list(row['antecedents'])
            consequent = list(row['consequents'])
            confidence = row['confidence']
            lift = row['lift']
            print(f"    {antecedent} → {consequent} (Conf: {confidence:.4f}, Lift: {lift:.4f})")
    print("-" * 100)

# Create summary DataFrame
summary_df = pd.DataFrame(results_summary)
print("\n" + "=" * 100)
print("SUMMARY TABLE: Confidence Threshold vs Number of Rules")
print("=" * 100)
print(summary_df.to_string(index=False))

print("\n\nKey Insights:")
print("1. As confidence threshold increases, the number of rules decreases")
print("2. Higher confidence threshold filters out weak rules")
print("3. Rules that survive higher thresholds are more reliable predictors")
print("4. There's a trade-off between coverage (number of rules) and quality (high confidence)")
```

## Lab 8 Summary and Conclusions

### Key Learnings:

1. **Frequent Itemset Mining (Apriori Algorithm)**
   - Efficiently finds all itemsets that appear in at least a given fraction of transactions
   - Support threshold acts as a filter for frequently co-occurring items
   - Higher support = fewer but more reliable patterns

2. **Association Rules**
   - Express relationships between items in the form: A → B
   - Three important metrics:
     - **Support**: Overall frequency of the rule
     - **Confidence**: Reliability of the rule (P(B|A))
     - **Lift**: Strength of association (how much better than random chance)

3. **Practical Implications**
   - In market basket analysis, discovering that "Customers who buy Milk often buy Eggs" (high confidence + high lift)
   - Can be used for product recommendations and cross-selling strategies
   - Threshold selection is crucial: too low = noise, too high = missing patterns

4. **Algorithm Comparison**
   - **Apriori**: Easy to understand, works well with smaller datasets
   - **FP-Growth**: More efficient for larger datasets, uses tree structure instead of candidate generation
