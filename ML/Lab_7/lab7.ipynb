# Lab 7: Association Rule Mining

## Objective
Explore Association Rule Mining with variations on different datasets using the Apriori algorithm and association rule generation.

### Theory Overview
- **Support**: The proportion of transactions that contain the itemset
- **Confidence**: Ratio of transactions containing both items to transactions containing the antecedent
- **Lift**: Ratio of observed support to expected support if items were independent
- **Apriori Algorithm**: Breadth-first search strategy for finding frequent itemsets


```python
# Import required libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
# Create the dataset
dataset = [
    ['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
    ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
    ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']
]

print("Dataset:")
for idx, transaction in enumerate(dataset, 1):
    print(f"Transaction {idx}: {transaction}")

# Convert transactions to one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

print("\nTransaction-Item Matrix (One-Hot Encoded):")
print(df)
```

## Task 1: Generate Frequent Itemsets with Minimum Support = 0.6

Using the Apriori algorithm to identify itemsets with support >= 0.6

```python
# Task 1: Apriori with minimum support = 0.6
min_support_1 = 0.6
frequent_itemsets_1 = apriori(df, min_support=min_support_1, use_colnames=True)

# Sort by support in descending order
frequent_itemsets_1 = frequent_itemsets_1.sort_values('support', ascending=False).reset_index(drop=True)

print(f"Frequent Itemsets with Minimum Support = {min_support_1}:")
print(f"Total number of frequent itemsets: {len(frequent_itemsets_1)}\n")

# Display itemsets with their support values
for idx, row in frequent_itemsets_1.iterrows():
    itemset = list(row['support'].index) if hasattr(row['support'], 'index') else list(row['itemsets'])
    support = row['support']
    print(f"Itemset {idx+1}: {itemset} -> Support: {support:.4f}")

# Create a more detailed display
detailed_df_1 = frequent_itemsets_1.copy()
detailed_df_1['itemset_list'] = detailed_df_1['itemsets'].apply(lambda x: list(x))
detailed_df_1['itemset_size'] = detailed_df_1['itemsets'].apply(lambda x: len(x))

print("\n" + "="*80)
print("Summary Table - Task 1:")
print("="*80)
display_df_1 = pd.DataFrame({
    'Itemset': detailed_df_1['itemset_list'],
    'Support': detailed_df_1['support'].round(4),
    'Itemset Size': detailed_df_1['itemset_size']
})
print(display_df_1.to_string(index=False))
```

## Task 2: Generate Frequent Itemsets with Minimum Support = 0.7 and Compare Results

Regenerate frequent itemsets with higher support threshold and compare with Task 1 results

```python
# Task 2: Apriori with minimum support = 0.7
min_support_2 = 0.7
frequent_itemsets_2 = apriori(df, min_support=min_support_2, use_colnames=True)

# Sort by support in descending order
frequent_itemsets_2 = frequent_itemsets_2.sort_values('support', ascending=False).reset_index(drop=True)

print(f"Frequent Itemsets with Minimum Support = {min_support_2}:")
print(f"Total number of frequent itemsets: {len(frequent_itemsets_2)}\n")

# Display itemsets with their support values
for idx, row in frequent_itemsets_2.iterrows():
    itemset = list(row['itemsets'])
    support = row['support']
    print(f"Itemset {idx+1}: {itemset} -> Support: {support:.4f}")

# Create a more detailed display
detailed_df_2 = frequent_itemsets_2.copy()
detailed_df_2['itemset_list'] = detailed_df_2['itemsets'].apply(lambda x: list(x))
detailed_df_2['itemset_size'] = detailed_df_2['itemsets'].apply(lambda x: len(x))

print("\n" + "="*80)
print("Summary Table - Task 2:")
print("="*80)
display_df_2 = pd.DataFrame({
    'Itemset': detailed_df_2['itemset_list'],
    'Support': detailed_df_2['support'].round(4),
    'Itemset Size': detailed_df_2['itemset_size']
})
print(display_df_2.to_string(index=False))
```

```python
# Comparison between Task 1 and Task 2
print("\n" + "="*80)
print("COMPARISON: Task 1 (min_support=0.6) vs Task 2 (min_support=0.7)")
print("="*80)

comparison_data = {
    'Metric': ['Minimum Support', 'Number of Frequent Itemsets', 'Max Support Found'],
    'Task 1 (support=0.6)': [
        f"{min_support_1:.2f}",
        len(frequent_itemsets_1),
        f"{frequent_itemsets_1['support'].max():.4f}"
    ],
    'Task 2 (support=0.7)': [
        f"{min_support_2:.2f}",
        len(frequent_itemsets_2),
        f"{frequent_itemsets_2['support'].max():.4f}"
    ]
}

comparison_df = pd.DataFrame(comparison_data)
print(comparison_df.to_string(index=False))

print("\n" + "-"*80)
print("Key Observations:")
print("-"*80)
print(f"1. When support threshold increased from {min_support_1} to {min_support_2}:")
print(f"   - Number of itemsets decreased from {len(frequent_itemsets_1)} to {len(frequent_itemsets_2)}")
print(f"   - Difference: {len(frequent_itemsets_1) - len(frequent_itemsets_2)} fewer itemsets")

print(f"\n2. Items removed when increasing support threshold to {min_support_2}:")
itemsets_1_set = set(map(frozenset, frequent_itemsets_1['itemsets']))
itemsets_2_set = set(map(frozenset, frequent_itemsets_2['itemsets']))
removed_itemsets = itemsets_1_set - itemsets_2_set
for itemset in removed_itemsets:
    print(f"   - {list(itemset)}")

print("\n3. This demonstrates the inverse relationship between support threshold and")
print("   the number of frequent itemsets generated by Apriori algorithm.")
```

## Task 3: Generate Association Rules with Confidence and Lift Values

Extract association rules from frequent itemsets with minimum confidence threshold

```python
# Task 3: Generate association rules from Task 1 frequent itemsets (support=0.6)
min_confidence = 0.6

rules = association_rules(frequent_itemsets_1, metric="confidence", min_threshold=min_confidence)

# Calculate additional metrics
if len(rules) > 0:
    # Sort by lift in descending order
    rules = rules.sort_values('lift', ascending=False).reset_index(drop=True)
    
    print(f"Association Rules (Minimum Confidence = {min_confidence}):")
    print(f"Total number of rules generated: {len(rules)}\n")
    
    # Display detailed rules
    for idx, row in rules.iterrows():
        antecedents = list(row['antecedents'])
        consequents = list(row['consequents'])
        confidence = row['confidence']
        lift = row['lift']
        support = row['support']
        
        print(f"Rule {idx+1}: {antecedents} => {consequents}")
        print(f"  Support: {support:.4f}")
        print(f"  Confidence: {confidence:.4f}")
        print(f"  Lift: {lift:.4f}")
        print()
    
    # Create a detailed summary table
    print("\n" + "="*100)
    print("Summary Table - Task 3 (Association Rules):")
    print("="*100)
    
    rules_display = pd.DataFrame({
        'Antecedent': rules['antecedents'].apply(lambda x: list(x)),
        'Consequent': rules['consequents'].apply(lambda x: list(x)),
        'Support': rules['support'].round(4),
        'Confidence': rules['confidence'].round(4),
        'Lift': rules['lift'].round(4)
    })
    
    print(rules_display.to_string(index=False))
    
else:
    print(f"No association rules found with minimum confidence = {min_confidence}")
    print("Try lowering the confidence threshold.")
```

## Task 4: Explore the Impact of Different Confidence Thresholds

Analyze how changing confidence thresholds affects the number and quality of association rules

```python
# Task 4: Explore different confidence thresholds
confidence_thresholds = [0.5, 0.6, 0.7, 0.8, 0.9]

print("="*100)
print("Impact of Different Confidence Thresholds on Association Rules")
print("="*100)
print("\nBase frequent itemsets: min_support = 0.6\n")

# Store results for comparison
threshold_results = []

for confidence in confidence_thresholds:
    temp_rules = association_rules(frequent_itemsets_1, metric="confidence", min_threshold=confidence)
    
    num_rules = len(temp_rules)
    if num_rules > 0:
        avg_confidence = temp_rules['confidence'].mean()
        avg_lift = temp_rules['lift'].mean()
        max_lift = temp_rules['lift'].max()
    else:
        avg_confidence = 0
        avg_lift = 0
        max_lift = 0
    
    threshold_results.append({
        'Confidence Threshold': f"{confidence:.1f}",
        'Number of Rules': num_rules,
        'Average Confidence': f"{avg_confidence:.4f}",
        'Average Lift': f"{avg_lift:.4f}",
        'Max Lift': f"{max_lift:.4f}"
    })
    
    print(f"Confidence Threshold: {confidence:.1f}")
    print(f"  - Number of rules generated: {num_rules}")
    if num_rules > 0:
        print(f"  - Average Confidence: {avg_confidence:.4f}")
        print(f"  - Average Lift: {avg_lift:.4f}")
        print(f"  - Maximum Lift: {max_lift:.4f}")
    print()

# Display comparison table
print("\n" + "="*100)
print("Summary Table - Task 4 (Confidence Threshold Impact):")
print("="*100)
threshold_df = pd.DataFrame(threshold_results)
print(threshold_df.to_string(index=False))

# Detailed analysis for specific thresholds
print("\n" + "-"*100)
print("Detailed Analysis:")
print("-"*100)

print("\nRules at Different Confidence Levels:\n")

# Show rules at 0.5 confidence
print("1. Rules at Confidence >= 0.5:")
rules_05 = association_rules(frequent_itemsets_1, metric="confidence", min_threshold=0.5)
if len(rules_05) > 0:
    for idx, row in rules_05.iterrows():
        print(f"   {list(row['antecedents'])} => {list(row['consequents'])} " +
              f"(confidence: {row['confidence']:.4f}, lift: {row['lift']:.4f})")
else:
    print("   No rules found")

print("\nRules at Confidence >= 0.7:")
rules_07 = association_rules(frequent_itemsets_1, metric="confidence", min_threshold=0.7)
if len(rules_07) > 0:
    for idx, row in rules_07.iterrows():
        print(f"   {list(row['antecedents'])} => {list(row['consequents'])} " +
              f"(confidence: {row['confidence']:.4f}, lift: {row['lift']:.4f})")
else:
    print("   No rules found")

print("\nRules at Confidence >= 0.9:")
rules_09 = association_rules(frequent_itemsets_1, metric="confidence", min_threshold=0.9)
if len(rules_09) > 0:
    for idx, row in rules_09.iterrows():
        print(f"   {list(row['antecedents'])} => {list(row['consequents'])} " +
              f"(confidence: {row['confidence']:.4f}, lift: {row['lift']:.4f})")
else:
    print("   No rules found")

print("\n" + "-"*100)
print("Key Insights:")
print("-"*100)
print("1. Higher confidence thresholds result in fewer but stronger association rules")
print("2. Rules with high lift indicate strong positive associations between items")
print("3. A lift > 1 suggests items are more likely to occur together than independently")
print("4. As confidence increases, we eliminate weaker rules, keeping only the strongest associations")
```

## Summary and Conclusions

### What We Learned:

1. **Apriori Algorithm**: Successfully identified frequent itemsets at different support thresholds
   - At support=0.6: More itemsets discovered (more lenient threshold)
   - At support=0.7: Fewer itemsets (stricter requirement)

2. **Association Rules**: Generated actionable insights about item relationships
   - **Support**: Shows how often itemsets appear in transactions
   - **Confidence**: Indicates rule reliability (if X then Y)
   - **Lift**: Measures association strength (lift > 1 = positive association)

3. **Confidence Threshold Impact**: 
   - Increasing confidence filters out weaker rules
   - Higher confidence = fewer but stronger rules
   - Optimal threshold depends on business requirements

In a real grocery store context, high-confidence, high-lift rules could identify:
- Products frequently bought together
- Product placement and bundling strategies
